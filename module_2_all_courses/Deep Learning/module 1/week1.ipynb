```python
# BackPropogation algorithm:

# at output layer:
# del5= yhat*(1-yhat)*(y-yhat)

# at hidden layer
# del3 : y3*(1-y3)*(w13*del5)
# del4 : y4*(1-y4)*(w14*del5)
# del_w13=learning_rate*del5*y3
# del_w14=lr*del5*y4
# w13=w13+del_w13
# w14=w14+del_w14

# at input layer
# del_w11=lr*del3*x1
# del_w12=lr*del4*x1
# del_w21=lr*del3*x2
# del_w22=lr*del4*x2
# w11=w11+del_w11
# w12=w12+del_w12
# w13=w13+del_w13
# w14=w14+del_w14



# What if we initialize the weights to zero:
# it causes all neurons in a given layer to compute the exact same function , preventing the netowrk from learning meaningfull patterns through backpropogation. This is also known as symmetric weights or dead neuron problem.
# Ways to initialize the weights:
# random initialization
# Xavier/Glorot initialization adjust the scale of random based on the number of input and output connections of the layer to maintain variance of activations through the network.

# He initialization: similiar to xavier but optimized for layers using the ReLU activation.


# we should initialize the weights with very small values.in case of tanh or sigmoid functions if the weights are too large when we compute the activation values some values of z will be either very large or very small. and we more likely end up at these fat parts of the tanh where the slope of the gradient is very small and gradient descent will be very slow.
#  
# w=np.random.randn((3,2))*0.01

# Hyperparameters:
# learning rate , number of iteration,number of hidden layers and nodes in hidden layers,choice of activation,momentum ,mini batch size,etc.







```

```python
import numpy as np
class NeuralNetwork:
    def __init__(self,input_size,hidden_size,output_size):
        self.input_size=input_size
        self.output_size=output_size
        self.hidden_size=hidden_size

        self.weights_input_hidden=np.random.randn(self.input_size,self.hidden_size)
        self.weights_hidden_output=np.random.randn(self.hidden_size,self.output_size)

        self.bias_hidden=np.zeros((1,self.hidden_size))
        self.bias_output=np.zeros((1,self.output_size))

    def sigmoid(self,x):
        return 1/(1+np.exp(-x))

    def sigmoid_derivative(self,x):
        return x*(1-x)


    def feedforward(self,X):
        self.hidden_activation=np.dot(X,self.weights_input_hidden)+self.bias_hidden
        self.hidden_output=self.sigmoid(self.hidden_activation)
        self.output_activation=np.dot(self.hidden_output,self.weights_hidden_output)+self.bias_output
        self.predict_output=self.sigmoid(self.output_activation)
        return self.predict_output

    def backward(self,X,y,lr):
        output_error=y-self.predict_output
        output_delta=output_error*self.sigmoid_derivative(self.predict_output)
        hidden_error=np.dot(output_delta,self.weights_hidden_output.T)
        hidden_delta=hidden_error*self.sigmoid(self.hidden_output)
        self.weights_hidden_output+=np.dot(self.hidden_output.T,output_delta)*lr
        self.bias_output+=np.sum(output_delta,axis=0,keepdims=True)*lr
        self.weights_input_hidden+=np.dot(X.T,hidden_delta)*lr
        self.bias_hidden+=np.sum(hidden_delta,axis=0,keepdims=True)*lr

    
    def train(self,X,y,epochs,lr):
        for epoch in range(epochs):
            output=self.feedforward(X)
            self.backward(X,y,lr)
            if epoch%40==0:
                loss=np.mean(np.square(y-output))
                print(f"Epoch {epoch}, Loss:{loss}")


        
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

nn=NeuralNetwork(input_size=3,hidden_size=4,output_size=1)
nn.train(X,y,epochs=1000,lr=0.001)
output=nn.feedforward(X)
print(output)
```
