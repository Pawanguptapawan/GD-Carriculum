# Convolutional Neural Networks (CNN) — Study Notes

---

## 1. Padding

When we convolve an image of size **N×N** with a **3×3** kernel, the result is **(N−2)×(N−2)**. This causes two problems:
- The output shrinks with every convolution.
- Edge/corner pixels get far less attention than central pixels.

**Solution — Padding:** Add extra border pixels around the image before convolution.

| Mode | Description |
|------|-------------|
| **Valid** | No padding; output shrinks |
| **Same** | Pad so output size = input size |

**Same padding formula:**

> p = (f − 1) / 2, where f = filter size

**Example:**
- Input: 6×6, Kernel: 3×3
- After padding (p=1): 8×8
- Output: 6×6 ✅

---

## 2. Stride Convolutions

**Stride (s)** — the number of pixels the kernel shifts across the input at each step.

**Output size formula:**

> Output = ⌊(N + 2p − f) / s + 1⌋

> **Note:** Cross-correlation flips the kernel both horizontally and vertically.

---

## 3. Convolution Over Volume (3D)

For a colour image:

| Item | Size |
|------|------|
| Input image | N × N × 3 |
| Kernel | n × n × 3 |
| Output (single filter) | (N−n+1) × (N−n+1) |

Multiple filters can be applied simultaneously. With **N'c** filters:

> Output = (N−f+1) × (N−f+1) × N'c

---

## 4. CNN Layer Notation

For layer **l**:

| Symbol | Meaning |
|--------|---------|
| f[l] | Filter size |
| p[l] | Padding |
| s[l] | Stride |
| n_c[l] | Number of filters |
| Weights | f[l] × f[l] × n_c[l−1] × n_c[l] |
| Bias | n_c[l] |

**Height/Width at each layer:**

> n_h[l] = ⌊(n_h[l−1] + 2·p[l] − f[l]) / s[l] + 1⌋

> A[l] = m × n_h[l] × n_w[l] × n_c[l]

---

## 5. Types of CNN Layers

1. **Convolution (CONV)** — Applies filters to learn spatial features.
2. **Pooling (POOL)** — Downsamples feature maps.
3. **Fully Connected (FC)** — Standard dense layer.

---

## 6. Pooling

Pooling reduces spatial dimensions to cut computation and prevent overfitting.

| Type | Operation |
|------|-----------|
| **Max Pooling** | Takes the highest value in each region |
| **Average Pooling** | Takes the average value in each region |

> Max Pooling has **no learnable parameters** — only hyperparameters f and s.

> Convention: (Conv + MaxPool) is often counted as **one layer** since MaxPool has no learnable parameters.

---

## 7. Example CNN Architecture

```
32×32×3
 → Conv1 (f=5, s=1, 6 filters)  → 28×28×6
 → MaxPool1 (f=2, s=2)          → 14×14×6
 → Conv2 (f=5, s=1, 10 filters) → 10×10×10
 → MaxPool2 (f=2, s=2)          → 5×5×10
 → Flatten                       → 250 units
 → FC                            → 120 units  [W:(120,250), b:(120)]
 → FC                            → 50 units   [W:(50,120),  b:(50)]
 → Softmax                       → Output
```

> As we go deeper: Height ↓, Width ↓, Channels ↑.

---

## 8. Why Convolutions Work Well

Two key properties keep parameter counts small:

- **Parameter Sharing** — A feature detector useful in one image region works in another too.
- **Sparsity of Connections** — Each output depends only on a small number of inputs.

---

## 9. Classic Networks

| Network | Notable Use |
|---------|------------|
| **LeNet-5** | Handwritten digit recognition |
| **AlexNet** | Large-scale image classification |
| **VGG-16** | Deep uniform architecture |

---

## 10. ResNet (Residual Networks)

**Problem:** Very deep networks suffer from the *vanishing gradient* problem.

**Solution — Skip Connections (Residual Blocks):**

Standard path:
```
a[l] → Linear → ReLU → a[l+1] → Linear → ReLU → a[l+2]
```

ResNet path (shortcut added before final ReLU):
```
a[l+2] = g(z[l+2] + a[l])
```

**Why it works:** If extra layers learn nothing (W=0, b=0), the identity function passes through unchanged — so adding layers can't hurt performance.

---

## 11. 1×1 Convolutions (Network in Network)

A 1×1 convolution acts like a fully connected layer applied per pixel — it can shrink or expand the channel depth with minimal spatial cost.

---

## 12. Inception Network

Applies **multiple filter sizes in parallel** (1×1, 3×3, 5×5, max pool) within a single layer ("Inception module"), then concatenates the results. This lets the network decide which filter size is most useful.

---

## 13. MobileNet

Designed for **low-cost deployment** on mobile/embedded devices.

### Standard Convolution (cost baseline)
```
6×6×3  →  3×3×3 filters (5 filters)  →  4×4×5
Cost: 3×3×3 × 5 × 4×4 = 2,160 multiplications
```

### Depthwise Separable Convolution

**Step 1 — Depthwise Convolution** (one filter per channel):
```
6×6×3  →  3×3 (3 filters, one per channel)  →  4×4×3
Cost: 3×3 × 4×4 × 3 = 432
```

**Step 2 — Pointwise Convolution** (1×1 to mix channels):
```
4×4×3  →  1×1×3 (5 filters)  →  4×4×5
Cost: 1×1×3 × 4×4 × 5 = 240
```

**Total: 672 vs 2,160 → ~31% of original cost (69% savings)**

### MobileNet V1
```
Input → (Depthwise + Pointwise) × 13 → Pooling → FC → Softmax → Output
```

### MobileNet V2 — Bottleneck Block
```
Input → [Expansion → Depthwise → Projection (Pointwise)] × 17 → Pooling → FC → Softmax
         |_______________________Residual Connection__________________________|
```
The **projection** step compresses channels back down, reducing memory footprint for the next block.

---

## 14. EfficientNet

Provides a principled way to **scale** a network (depth, width, resolution) up or down for a target device.

## 16. Sliding Window Detection

- Crop a window, resize to CNN input size, run CNN, shift and repeat.
- Repeat with larger windows.

**Disadvantage:** Computationally expensive — many redundant forward passes.

---

## 17. Convolutional Implementation of Sliding Windows

Convert FC layers to convolutional layers (using matching filter sizes). The entire image is processed in **one forward pass**; each spatial position in the output volume corresponds to one sliding window result.

**Example pipeline (FC → Conv):**
```
14×14×3
 → 5×5 (16 filters)     → 10×10×16
 → MaxPool (2×2)         → 5×5×16
 → 5×5×16 (400 filters)  → 1×1×400  [was FC]
 → 1×1 (400 filters)     → 1×1×400  [was FC]
 → 1×1 (4 filters)+Softmax → 1×1×4
```

---

## 18. YOLO (You Only Look Once)

- Divide image into a grid (e.g., 3×3).
- For each cell, predict: `y = [Pc, bx, by, bh, bw, c1, c2, c3]`.
- Output shape: **3×3×8** (for one anchor box, 3 classes).
- Uses a single convolutional forward pass over the full image.

---

## 19. Intersection over Union (IoU)

> IoU = Area of Intersection / Area of Union

A prediction is considered correct if **IoU ≥ 0.5**.

---

## 20. Non-Max Suppression (NMS)

Prevents multiple detections of the same object.

**Algorithm:**
1. Discard all boxes with Pc ≤ 0.6.
2. Pick the box with the highest Pc → output as prediction.
3. Discard any remaining box with IoU ≥ 0.5 against the chosen box.
4. Repeat until no boxes remain.

---

## 21. Anchor Boxes

Allows a single grid cell to detect **multiple objects**.

- Predefine k anchor box shapes.
- Each grid cell predicts one vector per anchor box.

**Label with 2 anchor boxes:**
```
y = [Pc, bx, by, bh, bw, c1, c2, c3,   ← Anchor box 1
     Pc, bx, by, bh, bw, c1, c2, c3]   ← Anchor box 2
```

Output shape: **3×3 × 2 × 8**

**Limitations:**
- Doesn't handle >k objects in one cell.
- Doesn't handle two objects with the same anchor box shape in one cell.

---

1. Divide image into 3×3 grid.

## 24. Semantic Segmentation with U-Net

- Assigns a class label to **every pixel** in the image.
- U-Net uses an encoder–decoder structure with **skip connections**.

**Architecture pattern:**
- **Encoder (contracting):** Convolutions + MaxPool → H ↓, W ↓, Channels ↑
- **Decoder (expanding):** Transpose convolutions → H ↑, W ↑, Channels ↓
- Skip connections copy encoder feature maps directly to the decoder.

---

## 25. Transpose Convolution

The reverse of a normal convolution — **upsamples** a small feature map to a larger one.

**Example:** 2×2 input → 3×3 filter (p=1, s=2) → 4×4 output

Instead of placing the filter over the input, the filter is placed over the **output** grid, with the input value scaling the filter weights.

# Face Recognition — Study Notes

---

## 1. One-Shot Learning

Traditional classifiers need many examples per class. Face recognition must work with **just one reference photo** per person.

> **One-shot learning** — Learn to recognise a person from a single example image.

Instead of training a softmax classifier (which would need retraining every time a new person is added), we learn a **similarity function** that generalises to new faces.

---

## 2. Similarity Function

> d(img1, img2) = degree of difference between two images

| Condition | Interpretation |
|-----------|---------------|
| d(img1, img2) ≤ τ (threshold) | Same person |
| d(img1, img2) > τ | Different person |

This approach scales naturally — adding a new employee only requires storing their photo, not retraining the model.

---

## 3. Siamese Network

A Siamese network encodes each image into a **fixed-length vector** using the same CNN with shared weights.

```
Image x(i)  →  CNN  →  f[x(i)]   (encoding vector)
Image x(j)  →  CNN  →  f[x(j)]   (encoding vector)
```

### Learning Objective

Train the network parameters so that:

| Pair | Goal |
|------|------|
| x(i), x(j) are the **same** person | `‖f[x(i)] − f[x(j)]‖²` is **small** |
| x(i), x(j) are **different** people | `‖f[x(i)] − f[x(j)]‖²` is **large** |

The encoding `f[x]` is the network's learned "face fingerprint" — a compact representation that captures identity.

---

## 4. Triplet Loss

Triplet loss trains the network using three images at a time:

| Image | Role | Description |
|-------|------|-------------|
| **A** | Anchor | Reference image |
| **P** | Positive | Different photo of the **same** person as A |
| **N** | Negative | Photo of a **different** person |

### Core Constraint

We want the anchor–positive distance to be smaller than the anchor–negative distance **by at least a margin α**:

```
‖f[A] − f[P]‖²  +  α  ≤  ‖f[A] − f[N]‖²
```

Rearranged:

```
‖f[A] − f[P]‖²  −  ‖f[A] − f[N]‖²  +  α  ≤  0
```

> **Why the margin α?** Without it, the trivial solution f[A] = f[P] = f[N] = 0 satisfies the constraint perfectly but learns nothing useful. α forces a meaningful gap between positive and negative pairs.

### Triplet Loss Function

```
L(A, P, N) = max( ‖f[A]−f[P]‖² − ‖f[A]−f[N]‖² + α ,  0 )
```

- If the constraint is satisfied with margin → loss = 0 (no penalty).
- If the constraint is violated → loss > 0 (penalise).

### Overall Cost Function

```
J = (1/m) · Σ L(A, P, N)
```

### Choosing Hard Triplets

Random triplets are too easy — the network quickly satisfies them without learning useful features. Instead, choose **hard triplets** where:

```
d(A, P)  ≈  d(A, N)
```

These are cases where the anchor–positive and anchor–negative distances are close, forcing the network to learn fine-grained distinctions.

---

## 5. Face Verification as Binary Classification

An alternative to triplet loss — frame face verification as a **supervised binary classification** problem.

### Setup

- Input: A pair of images (A, B).
- Label: `1` if same person, `0` if different.

### Model

```
ŷ = sigmoid( wᵀ · |f[A] − f[B]| + b )
```

Where the input features are derived from the two encodings. A common variant uses the **chi-squared similarity**:

```
χ² = ( f[x(i)] − f[x(j)] )² / ( f[x(i)] + f[x(j)] )
```

### Training

| Pair | Label |
|------|-------|
| Same person | y = 1 |
| Different person | y = 0 |

Train a binary classifier on labelled pairs. At inference time, pre-compute and store the encoding for each known person — only one forward pass per verification query is needed.

---

## 6. Siamese vs Binary Classification — Comparison

| Aspect | Siamese / Triplet Loss | Binary Classification |
|--------|----------------------|----------------------|
| Training signal | Triplets (A, P, N) | Labelled pairs (same / different) |
| Output | Distance between encodings | Probability of same identity |
| Scalability | Add new person with one photo | Add new person with one photo |
| Hard example mining | Required for best performance | Not required |
| Deployment | Compare encodings directly | Run classifier on encoding pairs |


# Neural Style Transfer — Study Notes

---

## 1. What is Neural Style Transfer?

Given two input images, generate a new image that combines:

- The **content** of a content image (C)
- The **style** of a style image (S)

```
Content Image (C)  +  Style Image (S)  →  Generated Image (G)
   (e.g. photo)        (e.g. painting)       (stylised photo)
```

A **pre-trained CNN** (e.g. VGG) is used as a fixed feature extractor — we never update its weights. Instead, we update the **pixels of the generated image G** directly via gradient descent.

---

## 2. Overall Cost Function

> J(G) = α · J_content(C, G) + β · J_style(S, G)

| Term | Role |
|------|------|
| `J_content(C, G)` | Measures how similar G's content is to C |
| `J_style(S, G)` | Measures how similar G's style is to S |
| `α` | Hyperparameter — weight on content |
| `β` | Hyperparameter — weight on style |

Increasing α/β emphasises content; decreasing it emphasises style.

---

## 3. Finding the Generated Image G

Unlike regular training, the **image pixels are the parameters** we optimise.

```
1. Initialise G randomly (random noise image)
2. Compute J(G) using the frozen pretrained CNN
3. Update pixels via gradient descent:

   G  ←  G  −  lr · ∂J/∂G

4. Repeat until convergence
```

---

## 4. Content Cost Function

Use a **middle hidden layer** `l` of the pretrained CNN to compare content. Shallow layers capture low-level features (edges); deeper layers capture high-level structure (objects, layout) — a middle layer balances both.

Let:
- `a[l](C)` = activations of layer `l` on the content image
- `a[l](G)` = activations of layer `l` on the generated image

> J_content(C, G) = ½ · ‖ a[l](C) − a[l](G) ‖²

If the activations are similar, the images share the same content. Gradient descent nudges G's pixels until this difference is minimised.

---

## 5. Style Cost Function

### What is "Style"?

Style is captured by measuring **how correlated the activations are across different channels** within a layer. If channel `c1` detects vertical stripes and channel `c2` detects orange colour, a high correlation between them means the style tends to have vertical orange stripes.

### Gram Matrix (Style Matrix)

The style of an image at layer `l` is represented by a **Gram matrix** G[l], of shape `n_c × n_c`:

**For the Style image:**

```
G[l][S][c1][c2] = Σᵢ Σⱼ  a[l][S][i,j,c1]  ·  a[l][S][i,j,c2]
```

**For the Generated image:**

```
G[l][G][c1][c2] = Σᵢ Σⱼ  a[l][G][i,j,c1]  ·  a[l][G][i,j,c2]
```

Where:
- `i`, `j` — spatial positions (height, width)
- `c1`, `c2` — channel indices
- Each entry measures the **co-activation** of channels c1 and c2 across all spatial positions

### Style Cost at Layer l

> J_style[l](S, G) = 1 / (n_H · n_W · n_c)² · Σ_{c1} Σ_{c2} ( G[l][S][c1,c2] − G[l][G][c1,c2] )²

The normalisation term `(n_H · n_W · n_c)²` scales the loss regardless of layer size.

### Total Style Cost (Across All Layers)

Using multiple layers captures style at different levels of abstraction (textures, colours, patterns):

> J_style(S, G) = Σ_l  λ[l] · J_style[l](S, G)

Where `λ[l]` is a per-layer weighting hyperparameter controlling how much each layer contributes to the style.

---

## 6. Full Algorithm Summary

```
Inputs:  Content image C,  Style image S
Output:  Generated image G

1.  Load pretrained CNN (e.g. VGG-19), freeze all weights
2.  Initialise G as random noise
3.  For each iteration:
      a. Forward pass C through CNN  →  extract a[l](C)
      b. Forward pass S through CNN  →  compute Gram matrices G[l][S]
      c. Forward pass G through CNN  →  extract a[l](G), compute G[l][G]
      d. Compute J_content(C, G) using content layer activations
      e. Compute J_style(S, G)   using Gram matrix differences
      f. Compute J(G) = α·J_content + β·J_style
      g. Update G  ←  G − lr · ∂J/∂G
4.  Output final G
```

---

## 7. Key Concepts at a Glance

| Concept | Details |
|---------|---------|
| What we optimise | Pixels of G (not CNN weights) |
| CNN role | Fixed feature extractor |
| Content layer | One middle layer (e.g. `conv4_2` in VGG) |
| Style layers | Multiple layers, each with a weight λ[l] |
| Gram matrix size | n_c × n_c (channels × channels) |
| α >> β | Generated image looks more like content |
| β >> α | Generated image looks more like style |
