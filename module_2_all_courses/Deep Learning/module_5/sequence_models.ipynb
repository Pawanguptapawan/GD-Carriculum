# Recurrent Neural Networks (RNN) — Study Notes

---

## 1. Why RNNs Instead of Standard Networks?

Standard feedforward networks fall short on sequential data for two reasons:

| Problem | Explanation |
|---------|-------------|
| **Variable-length I/O** | Inputs and outputs can differ in length across examples (e.g. sentences of different word counts) |
| **No feature sharing across positions** | A feature learned at position 1 of a sentence is not reused at position 5 — parameters don't generalise across the sequence |

RNNs solve both by processing inputs step-by-step while carrying a **hidden state** that accumulates context from previous steps.

---

## 2. How RNNs Work

At each time step `t`, the RNN:
1. Takes the **current input** `x[t]`
2. Combines it with the **previous hidden state** `a[t-1]`
3. Produces a **new hidden state** `a[t]` and an **output** `y[t]`

```
x[1] → [ RNN ] → y[1]
            ↓ a[1]
x[2] → [ RNN ] → y[2]
            ↓ a[2]
x[3] → [ RNN ] → y[3]
            ↓ a[3]
          ...
```

The same weights are shared across **all time steps** — this is what enables feature sharing across positions.

---

## 3. RNN Equations

### Hidden State Update

> a[t] = g( W_aa · a[t−1]  +  W_ax · x[t]  +  b_a )

### Output

> y[t] = g( W_ya · a[t]  +  b_y )

| Symbol | Meaning |
|--------|---------|
| `a[t]` | Hidden state at time step t |
| `x[t]` | Input at time step t |
| `W_aa` | Weight matrix — previous hidden state → current hidden state |
| `W_ax` | Weight matrix — current input → hidden state |
| `W_ya` | Weight matrix — hidden state → output |
| `b_a` | Bias for the hidden state |
| `b_y` | Bias for the output |
| `g` | Activation function (typically tanh or ReLU) |

---

## 4. Backpropagation Through Time (BPTT)

Standard backprop doesn't work for sequences — instead, we use **Backpropagation Through Time (BPTT)**:

1. **Unroll** the RNN across all time steps (like a very deep feedforward net).
2. Compute the loss at each time step.
3. Propagate gradients **backwards through every time step** to update shared weights.

### Challenges

| Issue | Description | Fix |
|-------|-------------|-----|
| **Vanishing gradients** | Gradients shrink as they travel back through many steps → early time steps get almost no learning signal | Use LSTM / GRU; use ReLU; gradient clipping |
| **Exploding gradients** | Gradients grow uncontrollably | Gradient clipping |
| **Computational cost** | Long sequences require unrolling over many steps | Truncated BPTT — limit the number of steps gradients are propagated back |

---

## 5. Types of RNN Architectures

| Type | Input | Output | Example Use Cases |
|------|-------|--------|-------------------|
| **Many-to-Many** | Sequence | Sequence | Machine translation, named entity recognition |
| **Many-to-One** | Sequence | Single value | Sentiment analysis, text classification |
| **One-to-Many** | Single value | Sequence | Image captioning, music generation |
| **One-to-One** | Single value | Single value | Standard feedforward (not truly RNN) |

### Visual Summary

```
Many-to-Many        Many-to-One         One-to-Many         One-to-One
x→□→x→□→x→□        x→□→x→□→x→□        x→□                 x→□
  ↓   ↓   ↓                    ↓           ↓→□→□→□             ↓
  y   y   y                    y           y  y  y             y
```

---

## 6. RNN Variants

| Architecture | Key Idea | Solves |
|-------------|----------|--------|
| **Vanilla RNN** | Basic hidden state recurrence | — |
| **LSTM** | Input, forget, output gates control memory cell | Vanishing gradients, long-term dependencies |
| **GRU** | Simplified gating (reset + update gates) | Vanishing gradients (lighter than LSTM) |

> LSTMs and GRUs were specifically designed to allow the network to **selectively retain or forget** information over long sequences — solving the core weakness of vanilla RNNs.

---

## 7. Key Takeaways

```
Sequential data          →  use RNNs (not standard feedforward nets)
Long-term dependencies   →  use LSTM or GRU (not vanilla RNN)
Variable-length I/O      →  handled naturally by RNN's step-by-step processing
Shared weights           →  same W used at every time step = feature sharing across positions
Training method          →  Backpropagation Through Time (BPTT)
```

# RNN — Loss Functions, Sampling & Vanishing Gradients

---

## 1. Loss and Cost Functions

### Loss Function (per example)

Cross-entropy loss for a single training example:

> L(y, ŷ) = − Σ  y · log(ŷ)

Measures how well the predicted probability distribution ŷ matches the true label y. A perfect prediction gives loss = 0.

### Cost Function (over entire dataset)

> J(W, b) = (1/m) · Σ  L(y[i], ŷ[i])

The average loss across all `m` training examples. Gradient descent minimises J by updating W and b.

---

## 2. Sampling a Sequence from a Trained RNN

After training, we can use the RNN to **generate** new sequences by feeding each output back as the next input.

### Process

```
1. Start with a seed input x[1] (or a random/zero vector)
        ↓
2. Forward pass → get output distribution ŷ[1]
        ↓
3. Sample a token from ŷ[1]  (e.g. a word or character)
        ↓
4. Feed sampled token as x[2]
        ↓
5. Repeat until stopping criterion is met
```

### Stopping Criteria
- Reaching a maximum sequence length.
- Sampling an end-of-sequence (EOS) token.

### Sampling Strategies

| Strategy | How it works | Trade-off |
|----------|-------------|-----------|
| **Greedy sampling** | Always pick the highest-probability token | Fast but repetitive / less diverse |
| **Random sampling** | Sample from the full probability distribution | More diverse but can be incoherent |
| **Beam search** | Keep top-k candidate sequences at each step | Better quality, more compute |
| **Temperature sampling** | Scale logits before softmax to control randomness | Low temp = conservative; high temp = creative |

---

## 3. Vanishing Gradients in RNNs

### The Problem

During **Backpropagation Through Time (BPTT)**, gradients are multiplied together across every time step. If those values are < 1, repeated multiplication causes them to shrink exponentially:

```
Gradient at step t  ∝  (W_aa)^(t steps)

If  |W_aa| < 1  →  gradient → 0   (vanishes)
If  |W_aa| > 1  →  gradient → ∞   (explodes)
```

For long sequences (e.g. 100+ words), early time steps receive **near-zero gradients** — the model simply cannot learn dependencies that span many steps.

### Why Vanilla RNNs Struggle

A sentence like:

> *"The cat, which sat on the mat every day, **was** hungry."*

requires connecting **"cat"** (early) to **"was"** (late). A vanilla RNN's gradient signal from "was" has all but vanished by the time it reaches "cat".

### Solutions

| Technique | How it helps |
|-----------|-------------|
| **LSTM** | Gates (input, forget, output) selectively preserve gradients across long spans |
| **GRU** | Lighter gating mechanism (reset + update gates); similar benefit to LSTM |
| **Gradient clipping** | Caps gradient magnitude to prevent exploding gradients |
| **ReLU activation** | Avoids saturation zone of tanh/sigmoid; reduces vanishing in some cases |
| **Truncated BPTT** | Limits how far back gradients are propagated; reduces compute cost |

> **Bottom line:** Basic RNNs are not well-suited for capturing long-term dependencies. LSTMs and GRUs were specifically designed to solve this.

---

## 4. Vanishing vs Exploding — Quick Comparison

| | Vanishing Gradients | Exploding Gradients |
|-|---------------------|---------------------|
| **Symptom** | Model stops learning; early layers unchanged | Training diverges; loss becomes NaN |
| **Cause** | Gradient < 1 multiplied many times | Gradient > 1 multiplied many times |
| **Fix** | LSTM / GRU / better init | Gradient clipping |
| **More common in** | Deep / long-sequence RNNs | Same |

# Gated Recurrent Unit (GRU) — Study Notes

---

## 1. Why GRU?

Vanilla RNNs struggle to retain information across many time steps due to vanishing gradients. GRU solves this by introducing **gating mechanisms** that explicitly control what to remember and what to update at each step.

> GRU is a simplified but highly effective alternative to LSTM for capturing long-range dependencies.

---

## 2. Core Idea: Memory Cell

GRU introduces a **memory cell** `c<t>` that persists information across time steps.

> In GRU: `c<t> = a<t>` — the memory cell and the hidden state are the **same vector**.  
> (In LSTM these are separate — a key architectural difference.)

---

## 3. GRU Gates

GRU uses two gates, each outputting a value between 0 and 1 via the sigmoid function:

| Gate | Symbol | Formula | Role |
|------|--------|---------|------|
| **Update gate** | γ_u | σ( W_u · [c\<t−1\>, x\<t\>] + b_u ) | How much of the old memory to keep |
| **Reset gate** | γ_r | σ( W_r · [c\<t−1\>, x\<t\>] + b_r ) | How much of the past to use when computing the candidate |

---

## 4. Full GRU Equations (Step by Step)

### Step 1 — Reset Gate
> γ_r = σ( W_r · [c\<t−1\>, x\<t\>] + b_r )

Controls how much of the previous memory cell to expose when computing the candidate state.

### Step 2 — Candidate Memory Cell
> c̃\<t\> = tanh( W_c · [γ_r ⊙ c\<t−1\>,  x\<t\>] + b_c )

`γ_r ⊙ c<t−1>` gates how much past context flows into the new candidate. If γ_r ≈ 0, the candidate ignores the past entirely.

### Step 3 — Update Gate
> γ_u = σ( W_u · [c\<t−1\>, x\<t\>] + b_u )

Decides how much of the **old memory** to carry forward vs. how much of the **new candidate** to write in.

### Step 4 — Final Memory Cell Update
> c\<t\> = γ_u ⊙ c̃\<t\>  +  (1 − γ_u) ⊙ c\<t−1\>

| γ_u value | Effect |
|-----------|--------|
| γ_u ≈ 1 | Overwrite memory with new candidate |
| γ_u ≈ 0 | Keep previous memory unchanged → **long-term retention** |

> When γ_u ≈ 0: `c<t> ≈ c<t−1>` — the memory cell is preserved perfectly across many steps with no gradient degradation. This is the key mechanism that defeats the vanishing gradient problem.

---

## 5. Visualising the GRU Cell

```
 c<t−1> ──┬──────────────────────────────────────┐
           │                                      │
           │   ┌─ Reset Gate (γ_r) ──────────┐   │
           │   │  γ_r = σ(Wr·[c<t−1>, x<t>]) │   │
           │   └──────────────────────────────┘   │
           │            ↓  γ_r ⊙ c<t−1>           │
           │   ┌─ Candidate (c̃) ──────────────┐   │
 x<t>  ────┼──▶│  c̃ = tanh(Wc·[γ_r⊙c, x])   │   │
           │   └──────────────────────────────┘   │
           │            ↓  c̃<t>                   │
           │   ┌─ Update Gate (γ_u) ──────────┐   │
           │   │  γ_u = σ(Wu·[c<t−1>, x<t>]) │   │
           │   └──────────────────────────────┘   │
           │            ↓                         │
           └──▶  c<t> = γ_u⊙c̃ + (1−γ_u)⊙c<t−1> ─┘
                           ↓
                         a<t> = c<t>  →  output y<t>
```

---

## 6. Dimensionality

All vectors scale together. If the hidden state is 100-dimensional:

| Variable | Dimension |
|----------|-----------|
| `c<t>` | 100-d |
| `c̃<t>` | 100-d |
| `γ_u` | 100-d |
| `γ_r` | 100-d |

The `⊙` (element-wise / Hadamard) product means each **dimension has its own gate value** — the model can independently decide which memory dimensions to update or retain.

> Think of the 100-d gate vector as 100 individual on/off switches, each controlling one dimension of memory.

---

## 7. GRU vs Vanilla RNN

| Aspect | Vanilla RNN | GRU |
|--------|------------|-----|
| Memory cell | No (a\<t\> only) | Yes (c\<t\> = a\<t\>) |
| Gates | None | Update (γ_u) + Reset (γ_r) |
| Long-term dependencies | ❌ Poor | ✅ Good |
| Vanishing gradients | ❌ Severe | ✅ Mitigated |
| Parameters | Fewer | More (but fewer than LSTM) |

---

## 8. GRU vs LSTM

| Aspect | GRU | LSTM |
|--------|-----|------|
| Memory cell | c\<t\> = a\<t\> | c\<t\> ≠ a\<t\> (separate) |
| Gates | 2 (update, reset) | 3 (input, forget, output) |
| Complexity | Simpler | More expressive |
| Speed | Faster to train | Slower |
| Performance | Comparable on many tasks | Slightly better on some tasks |

# Bidirectional RNN (BRNN) — Study Notes

---

## 1. Motivation

A standard (unidirectional) RNN at time step `t` can only see inputs from the **past** — `x<1>` through `x<t>`. For many tasks, future context matters just as much.

**Example:**

> *"He said, 'Teddy Roosevelt was a great president.'"*  
> *"He said, 'Teddy bears are on sale!'"*

To correctly classify "Teddy" as a person's name vs. a toy, the model needs to see the **words that come after** it — impossible with a unidirectional RNN.

A Bidirectional RNN solves this by reading the sequence in **both directions simultaneously**.

---

## 2. Architecture

Two independent RNNs run in parallel on the same input sequence:

```
Forward RNN  →  a→<1>  →  a→<2>  →  a→<3>  →  a→<4>
                  ↑          ↑          ↑          ↑
               x<1>       x<2>       x<3>       x<4>

Backward RNN →  a←<1>  ←  a←<2>  ←  a←<3>  ←  a←<4>
                  ↑          ↑          ↑          ↑
               x<1>       x<2>       x<3>       x<4>
```

At each time step `t`, the outputs of both RNNs are **combined** (typically concatenated) to make a prediction:

> ŷ\<t\> = g( W_y · [ a→\<t\>,  a←\<t\> ]  +  b_y )

| Component | Sees |
|-----------|------|
| Forward RNN (a→) | x\<1\> … x\<t\> (past + present) |
| Backward RNN (a←) | x\<T\> … x\<t\> (future + present) |
| Combined output | **Full sequence context** at every step |

---

## 3. How It Works — Step by Step

```
1. Run the Forward RNN left → right, computing a→<1>, a→<2>, ..., a→<T>
2. Run the Backward RNN right → left, computing a←<T>, a←<T−1>, ..., a←<1>
3. At each time step t, concatenate [a→<t>, a←<t>]
4. Pass through output layer to get ŷ<t>
```

---

## 4. Combining the Two Directions

| Method | Description |
|--------|-------------|
| **Concatenation** (most common) | [a→\<t\>, a←\<t\>] — doubles the hidden dimension |
| **Summation** | a→\<t\> + a←\<t\> — keeps same dimension |
| **Average** | (a→\<t\> + a←\<t\>) / 2 |

---

## 5. Advantages

- Captures **both past and future context** at every position.
- Particularly powerful for NLP tasks: Named Entity Recognition, machine translation, sentiment analysis.
- Works with any RNN cell type — vanilla RNN, GRU, or LSTM cells can all be used inside a BRNN.

---

## 6. Limitations

| Limitation | Detail |
|-----------|--------|
| **Needs the full sequence upfront** | The backward RNN cannot start until it has seen the last token — real-time / streaming predictions are not possible |
| **Higher compute cost** | Roughly 2× the parameters and computation of a unidirectional RNN |
| **Not suitable for online tasks** | E.g. live speech recognition where words arrive one at a time |

> For most **offline NLP tasks** (translation, classification, NER) where the entire input is available at once, BRNNs are highly effective and widely used.

---

## 7. BRNN vs Unidirectional RNN

| Aspect | Unidirectional RNN | Bidirectional RNN |
|--------|--------------------|-------------------|
| Context used | Past only | Past + Future |
| Requires full sequence | No | Yes |
| NLP performance | Good | Better (on most tasks) |
| Real-time inference | ✅ Possible | ❌ Not possible |
| Parameters | Fewer | ~2× more |

## 4. Properties of Word Embeddings

### Similarity
Words with similar meanings have similar embedding vectors — their dot product or cosine similarity is high.

```
sim( embed("doctor"),  embed("physician") )  →  high
sim( embed("doctor"),  embed("banana")    )  →  low
```

### Compositionality (Analogical Reasoning)

Embeddings capture semantic relationships as **vector arithmetic**:

```
embed("king") − embed("man") + embed("woman")  ≈  embed("queen")
```

This means the direction from "man" → "king" (royalty) is the same as "woman" → "queen".

**More examples:**
```
embed("Paris") − embed("France") + embed("Italy")  ≈  embed("Rome")
embed("bigger") − embed("big")   + embed("small")  ≈  embed("smaller")
```

### Cosine Similarity Formula

Used to find the word `w` whose embedding best matches a target vector `v`:

> sim(v, embed(w)) = (v · embed(w)) / (‖v‖ · ‖embed(w)‖)

Returns a value between −1 (opposite) and 1 (identical direction).

```python
# Finding the analogy answer in code
import numpy as np

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# "king - man + woman = ?"
target = embed["king"] - embed["man"] + embed["woman"]
best_word = max(vocabulary, key=lambda w: cosine_similarity(embed[w], target))
# best_word → "queen"
```

## 5. Embedding Matrix

The embedding matrix **E** is a learnable parameter of shape **(vocab_size × embedding_dim)**.

```
vocab_size    = number of unique words (e.g. 10,000)
embedding_dim = size of each embedding vector (e.g. 300)

E  →  shape: (10,000 × 300)
```

Each **row** of E is the embedding for one word in the vocabulary.

### Looking Up an Embedding

To get the embedding for word at index `i`, multiply the one-hot vector by E:

```
embed(word_i) = E · one_hot(i)    →  shape: (300,)
```

In practice, this is implemented as a simple **index lookup** (not actual matrix multiplication) for efficiency.

```python
import torch.nn as nn

embedding_layer = nn.Embedding(num_embeddings=10000, embedding_dim=300)

# Lookup word at index 42
word_index = torch.tensor([42])
vector = embedding_layer(word_index)   # shape: (1, 300)
```

### How It Learns

- Initialised with random values at the start of training.
- Updated via backpropagation like any other weight matrix.
- After training, similar words end up with similar rows in E.

# Skip-Gram & Negative Sampling — Study Notes

---

## 1. Skip-Gram Model

### Objective

Given a **target (context) word**, predict the **surrounding words** within a window.

> This is the opposite of CBOW (Continuous Bag of Words), which predicts the target word from its surrounding context.

| Model | Input | Predicts |
|-------|-------|---------|
| **Skip-Gram** | One target word | Surrounding context words |
| **CBOW** | Surrounding context words | One target word |

### How It Works

For each word in the sentence, treat it as the **centre word** and create training pairs with words in a ±window around it.

**Example sentence:**
```
"I want a glass of orange juice to go along with my cereal."
```

With `orange` as the context word and window size = ±10:

| Context Word | Target Word |
|-------------|------------|
| orange | juice |
| orange | glass |
| orange | my |
| orange | of |

This becomes a **supervised learning problem** — given the context word, predict which word appears nearby.

### Training Signal

```
Input  : one-hot vector of context word (e.g. "orange")
         ↓
Embedding matrix E  →  300-d vector
         ↓
Softmax over entire vocabulary
         ↓
Output : probability distribution over all words
         (maximise P("juice" | "orange"))
```

---

## 2. Skip-Gram Limitation — Softmax Bottleneck

The standard Skip-Gram uses a **softmax over the full vocabulary** at every training step:

> P(t | c) = e^(θ_t · e_c) / Σ_{j=1}^{V} e^(θ_j · e_c)

For a vocabulary of **10,000–1,000,000 words**, computing this denominator at every step is extremely expensive.

### Solutions

| Solution | How it works |
|---------|-------------|
| **Hierarchical Softmax** | Arranges vocabulary in a binary tree; traverses O(log V) nodes instead of V |
| **Negative Sampling** | Reformulates as binary classification; only updates a small subset of weights per step ✅ |

---

## 3. Context Word Sampling Problem

When sampling context words randomly from a corpus, **very frequent words dominate**:

```
High frequency (sampled too often):   "the", "of", "a", "and", "to"
Low frequency (sampled too rarely):   "orange", "juice", "glass"
```

Frequent words carry less semantic information but would flood the training pairs if sampled uniformly.

**Fix — Subsampling:** Each word `w` is discarded from training with probability:

> P(discard) = 1 − √(t / f(w))

Where `f(w)` is the word's frequency and `t` is a threshold (typically 10⁻⁵). High-frequency words are more likely to be dropped, balancing the training signal.

---

## 4. Negative Sampling

### Core Idea

Instead of the expensive full-vocabulary softmax, **reframe the problem as binary classification**:

> *"Does this (context, target) pair come from the real data, or is it a random fake pair?"*

### Building the Training Data

For each real (context, target) pair → **1 positive example** + **k negative examples**:

| Context | Word | Label |
|---------|------|-------|
| orange | juice | 1 ✅ (real pair) |
| orange | king | 0 ❌ (random word) |
| orange | book | 0 ❌ (random word) |
| orange | the | 0 ❌ (random word) |
| orange | cat | 0 ❌ (random word) |

Typical values of **k** (number of negatives):
- Small datasets → k = 5–20
- Large datasets → k = 2–5

### The Binary Classifier

For each pair (c, t), learn:

> P(y=1 | c, t) = σ( θ_t · e_c )

Just a **sigmoid** — no expensive softmax sum over all words.

### What Gets Updated Per Step

```
Standard Softmax:   update weights for ALL V words  (e.g. 10,000)
Negative Sampling:  update weights for  k+1  words  (e.g. 5)
```

This makes training **~2,000× more efficient** for a 10k vocabulary.

---

## 5. Selecting Negative Samples

Negative words are **not** sampled uniformly — that would oversample common words like "the".

Instead, sample according to a **smoothed unigram distribution**:

> P(w) = f(w)^(3/4) / Σ_j f(j)^(3/4)

The 3/4 power smoothing:
- Increases the probability of rare words being selected
- Decreases the dominance of very frequent words
- Better balance than uniform or raw-frequency sampling

**Example:**
```
f("the")    = 0.10  →  0.10^0.75 = 0.056   (reduced)
f("orange") = 0.01  →  0.01^0.75 = 0.018   (boosted relative to its frequency)
```

---

## 6. Skip-Gram vs Negative Sampling — Full Comparison

| Aspect | Skip-Gram (vanilla) | Skip-Gram + Negative Sampling |
|--------|--------------------|-----------------------------|
| Output layer | Softmax over V words | Sigmoid per (c, t) pair |
| Weights updated per step | All V | k + 1 |
| Computational cost | O(V) | O(k) |
| Training speed | Slow for large vocab | ✅ Fast |
| Quality of embeddings | Good | ✅ Equally good |

---

## 7. Summary

```
Skip-Gram goal:
  Given centre word → predict surrounding words

Problem 1 — Softmax is slow:
  → Fix: Negative Sampling (binary classifier instead)

Problem 2 — Frequent words dominate:
  → Fix: Subsampling + smoothed unigram sampling for negatives

Result:
  Fast, scalable training of high-quality word embeddings
  with far fewer weight updates per step
```

# GloVe — Global Vectors for Word Representation

---

## 1. What is GloVe?

**GloVe (Global Vectors)** is a word embedding algorithm that learns vector representations by leveraging **global word co-occurrence statistics** from the entire corpus — rather than local context windows like Skip-Gram.

> Core idea: Two words that frequently appear near each other should have similar embeddings.

---

## 2. How GloVe Differs from Skip-Gram

| Aspect | Skip-Gram | GloVe |
|--------|----------|-------|
| Approach | Predict context words from target | Factorise co-occurrence matrix |
| Uses | Local sliding window | Global corpus statistics |
| Training pairs | (context, target) prediction | Co-occurrence counts |
| Efficiency | Negative sampling needed | Direct matrix optimisation |

---

## 3. Co-occurrence Matrix

GloVe starts by building a **co-occurrence matrix X** from the entire corpus.

> X[i, j] = number of times word **j** appears in the context of word **i**

| Symbol | Role |
|--------|------|
| i | Context word |
| j | Target word |
| X[i, j] | Co-occurrence count |

### Example

**Sentence:** *"I want a glass of orange juice to go along with my cereal."*

For the context word **"orange"**:

| j (target word) | X["orange", j] |
|----------------|---------------|
| juice | high count |
| glass | some count |
| cereal | some count |
| the | low count (frequent but unrelated) |
| king | 0 (never co-occurs) |

> GloVe only trains on pairs where **X[i, j] > 0** — pairs that have co-occurred at least once.

---

## 4. GloVe Objective Function

GloVe learns vectors by minimising the difference between the **dot product of two word vectors** and the **log of their co-occurrence count**:

> Minimise: Σ_{i,j} f(X[i,j]) · ( θᵢᵀ eⱼ + bᵢ + bⱼ − log X[i,j] )²

| Term | Meaning |
|------|---------|
| θᵢ | Context word vector for word i |
| eⱼ | Target word vector for word j |
| bᵢ, bⱼ | Bias terms |
| f(X[i,j]) | Weighting function — reduces influence of very frequent pairs |

### Weighting Function f(x)

```
f(x) = (x / x_max)^α    if x < x_max
f(x) = 1                 if x ≥ x_max
```

- Very frequent pairs (e.g. "the" + "of") are down-weighted so they don't dominate.
- Rare pairs still contribute but with less weight.
- Typically: x_max = 100, α = 3/4

---

## 5. What GloVe Learns

After training, the vectors θᵢ and eⱼ capture semantic relationships:

```
embed("king") − embed("man") + embed("woman")  ≈  embed("queen")
embed("Paris") − embed("France") + embed("Italy")  ≈  embed("Rome")
```

The final embedding for a word is typically the **average** of its context and target vectors:

> e_word = (θ_word + e_word) / 2

This is because in GloVe, i and j are **symmetric** — word i can play either context or target role.

---

## 6. GloVe vs Skip-Gram vs CBOW

| Property | CBOW | Skip-Gram | GloVe |
|----------|------|----------|-------|
| Training signal | Context → target | Target → context | Co-occurrence counts |
| Scope | Local window | Local window | **Global corpus** |
| Speed | Fast | Medium | Fast (after matrix build) |
| Rare words | Poor | Medium | Medium |
| Interpretability | Low | Medium | ✅ Higher |
| Analogy tasks | Good | Good | ✅ Very good |

---

## 7. Key Properties of GloVe Vectors

### Symmetry
Unlike Skip-Gram where context and target roles are distinct, in GloVe:

> X[i, j]  ≈  X[j, i]

Both roles contribute equally, making the embeddings more symmetric and stable.

### Additive Compositionality
GloVe embeddings support the same vector arithmetic as Word2Vec:

```
"bottle of water"  →  embed("bottle") + embed("of") + embed("water")
```

### Linear Substructure
Relationships between words form consistent **linear directions** in the embedding space:

```
Direction of "gender":   embed("man") → embed("woman")
                         embed("king") → embed("queen")   (same direction)

Direction of "capital":  embed("France") → embed("Paris")
                         embed("Italy")  → embed("Rome")  (same direction)
```

---

## 8. Summary

```
Corpus
  ↓
Build co-occurrence matrix X[i,j]        (count how often i and j appear together)
  ↓
Learn vectors θ and e                    (minimise weighted squared error vs log counts)
  ↓
Final embedding = (θ + e) / 2            (average context + target vectors)
  ↓
Rich semantic embeddings ready for NLP tasks
```

> GloVe's advantage: it uses the **entire corpus statistics at once** rather than learning from individual windows — giving it a more global, stable view of word relationships.

# BLEU Score — Study Notes

---

## 1. The Problem — Evaluating Multiple Valid Outputs

In tasks like **machine translation**, there are often many correct answers. How do you automatically measure accuracy when any of several reference translations could be correct?

> **BLEU (Bilingual Evaluation Understudy)** — measures how much the machine-generated output overlaps with one or more human reference translations.

---

## 2. Naive Precision — Why It Fails

**Setup:**

```
Ref 1    : "The cat is on the mat."       ("the" appears 2 times)
Ref 2    : "There is a cat on the mat."   ("the" appears 1 time)
MT output: "the the the the the the the"
```

**Naive precision** = words in output that appear in any reference / total output words:

> 7 / 7 = **1.0** (100%) ← clearly wrong

Every word "the" appears in a reference, so naive precision gives a perfect score to a nonsensical output.

---

## 3. Modified Precision (Unigrams)

**Fix:** Cap each word's count at the **maximum times it appears in any single reference**.

```
"the" appears max 2 times across references  →  clip count at 2

Modified precision = 2 / 7 ≈ 0.286
```

> **CountClip** = min(Count in output,  max Count in any reference)

This prevents the model from gaming the score by repeating high-frequency words.

---

## 4. Bigram BLEU — Example

A unigram score misses word order. **Bigrams** (pairs of consecutive words) capture local fluency.

**Setup:**

```
Ref 1    : "The cat is on the mat."
Ref 2    : "There is a cat on the mat."
MT output: "the cat the cat on the mat."
```

**Bigram counts:**

| Bigram | Count (output) | CountClip (max in any ref) |
|--------|---------------|--------------------------|
| the cat | 2 | 1 |
| cat the | 1 | 0 |
| cat on | 1 | 1 |
| on the | 1 | 1 |
| the mat | 1 | 1 |
| **Total** | **6** | **4** |

> Bigram precision = CountClip total / Count total = **4 / 6 ≈ 0.667**

---

## 5. General n-gram BLEU Formula

Extend the same idea to any n-gram size:

> Pₙ = Σ (CountClip of each n-gram) / Σ (Count of each n-gram)

| n | What it measures |
|---|-----------------|
| 1 (unigram) | Correct individual words |
| 2 (bigram) | Correct word pairs — local fluency |
| 3 (trigram) | Correct 3-word sequences |
| 4 (4-gram) | Longer phrase matches |

---

## 6. Combined BLEU Score

In practice, BLEU combines **P1 through P4** with a **Brevity Penalty (BP)** to penalise outputs that are too short:

> BLEU = BP · exp( (1/4) · Σₙ log Pₙ )

### Brevity Penalty

```
BP = 1                          if output length ≥ reference length
BP = exp(1 − ref_len / out_len) if output length < reference length
```

Without BP, a model could achieve high precision by outputting only one highly accurate word.

---

## 7. Worked Summary

```
MT output: "the cat the cat on the mat."

Unigram P1:
  "the"(2), "cat"(2), "on"(1), "mat"(1) → CountClip = 2+1+1+1 = 5
  Total count = 7
  P1 = 5/7 ≈ 0.714

Bigram P2:
  CountClip = 4,  Count = 6
  P2 = 4/6 ≈ 0.667

Combined (simplified, no BP):
  BLEU ≈ exp( (log 0.714 + log 0.667) / 2 ) ≈ 0.69
```

---

## 8. BLEU Strengths and Limitations

| Strength | Limitation |
|---------|-----------|
| Fast, automatic, language-independent | Doesn't capture meaning — only surface overlap |
| Correlates well with human judgement on average | A paraphrase with different words scores low even if correct |
| Handles multiple references naturally | Doesn't penalise grammatically wrong but word-correct output |
| Industry standard for MT evaluation | Not ideal for tasks needing creativity or diversity |

---

## 9. Quick Reference

```
Naive precision         →  count / total  (gameable with repeated words)
Modified precision      →  CountClip / total  (capped by reference frequency)
Unigram BLEU (P1)       →  word-level overlap
Bigram BLEU (P2)        →  phrase-level overlap
Combined BLEU (P1–P4)   →  with Brevity Penalty for short outputs
```

# Attention Model — Study Notes

---

## 1. Limitation of Traditional Encoder-Decoder

The standard Encoder-Decoder architecture uses two RNNs:

```
Input sentence
      ↓
 [ Encoder RNN ]  →  single fixed-length vector  →  [ Decoder RNN ]  →  Output sentence
```

| Sentence Length | Performance |
|----------------|-------------|
| Short | ✅ Works well |
| Long | ❌ Degrades significantly |

**Why it fails on long sentences:**
The encoder must compress the **entire input** into one fixed-length vector. For long sentences, this bottleneck loses information — the decoder has no way to recover details that didn't fit into the compressed representation.

> It's like asking someone to memorise a full paragraph word-for-word, then recite a translation from memory — works for short phrases, breaks down for long passages.

---

## 2. The Attention Idea

Instead of reading the whole sentence once and compressing it, the Attention Model lets the decoder **look back at the input at every step** — focusing on the most relevant parts for each word it generates.

> Attention mimics how human translators work: they don't memorise the full sentence; they glance at specific parts while writing each word of the translation.

```
Without attention:  encode everything → decode from single vector
With attention:     decode one word at a time, looking at relevant input words each step
```

---

## 3. Architecture Overview

### Encoder — Bidirectional RNN

The input sentence is encoded using a **Bidirectional RNN**, producing a rich feature vector (called an **annotation**) for each input word that captures both past and future context:

```
Input:   x⟨1⟩   x⟨2⟩   x⟨3⟩  ...  x⟨Tx⟩

Forward  →  a→⟨1⟩ →  a→⟨2⟩ →  a→⟨3⟩ → ...
Backward ←  a←⟨1⟩ ←  a←⟨2⟩ ←  a←⟨3⟩ ← ...

Annotation:  a⟨t⟩ = [ a→⟨t⟩ , a←⟨t⟩ ]   (concatenated)
```

Each annotation `a⟨t⟩` summarises the input word at position t with full bidirectional context.

---

### Decoder — Attention-Weighted RNN

At each decoding step `t'`, the decoder:

1. Computes **attention weights** α⟨t', t⟩ for every input position t
2. Builds a **context vector** c⟨t'⟩ as a weighted sum of annotations
3. Generates the output word using the context vector + previous hidden state

```
Step 1 — Attention weights:
  α⟨t', t⟩  =  how much focus to place on input word t
               when generating output word t'

  Σ_t α⟨t', t⟩ = 1   (weights sum to 1 via softmax)

Step 2 — Context vector:
  c⟨t'⟩  =  Σ_t  α⟨t', t⟩ · a⟨t⟩

Step 3 — Generate output:
  s⟨t'⟩  =  f( s⟨t'-1⟩ , y⟨t'-1⟩ , c⟨t'⟩ )
  y⟨t'⟩  =  softmax( s⟨t'⟩ )
```

---

## 4. Attention Weights — How They Are Computed

The attention weight α⟨t', t⟩ is computed in two stages:

```
Stage 1 — Alignment score (e):
  e⟨t', t⟩  =  a( s⟨t'-1⟩ , a⟨t⟩ )
  (a small neural network that scores how well input t matches decoding step t')

Stage 2 — Normalise with softmax:
  α⟨t', t⟩  =  exp(e⟨t', t⟩) / Σ_t exp(e⟨t', t⟩)
```

The alignment network `a` is trained jointly with the rest of the model — no separate supervision is needed.

---

## 5. Full Attention Mechanism — Step by Step

```
For each output word t' = 1, 2, ..., Ty:

  1. Take previous decoder hidden state  s⟨t'-1⟩
  2. Score each encoder annotation  a⟨t⟩  →  e⟨t', t⟩
  3. Softmax scores  →  attention weights  α⟨t', t⟩
  4. Weighted sum of annotations  →  context vector  c⟨t'⟩
  5. Decoder RNN produces  s⟨t'⟩  from  s⟨t'-1⟩, y⟨t'-1⟩, c⟨t'⟩
  6. Output layer produces  y⟨t'⟩  (translated word)
  7. Repeat for next output word
```

---

## 6. Visualising Attention Weights

Attention weights can be plotted as a **heatmap** (input words × output words). Bright cells show which input words the model focused on for each output word.

```
              je    ne    suis   pas    étudiant
              ↓     ↓     ↓      ↓      ↓
I          [ 0.9  0.0   0.0    0.0    0.0  ]
am         [ 0.0  0.8   0.1    0.0    0.0  ]
not        [ 0.0  0.1   0.1    0.8    0.0  ]
a student  [ 0.0  0.0   0.1    0.1    0.9  ]
```

> This heatmap reveals that the model has learned alignment between languages — even when word order differs.

---

## 7. Why Attention Solves the Long Sentence Problem

| Issue | Encoder-Decoder | Attention Model |
|-------|----------------|----------------|
| Information storage | All info → one vector | Info kept in all annotations |
| Long sentence handling | ❌ Degrades | ✅ Maintained |
| Word alignment | ❌ Implicit | ✅ Explicit via α weights |
| Interpretability | ❌ Black box | ✅ Attention heatmap |
| Computational cost | O(1) context | O(Tx × Ty) — quadratic |

> The one tradeoff: attention is **quadratic** in sequence length (every output step attends to every input position). For very long sequences this becomes expensive — solved later by **Transformers** with efficient attention mechanisms.

---

## 8. Summary

```
Traditional Encoder-Decoder:
  Entire sentence → single vector → decode (struggles with long inputs)

Attention Model:
  Bidirectional RNN encodes each word → annotation vectors
  At each decode step:
    → compute α weights (what to focus on)
    → build context vector (weighted sum of annotations)
    → generate next word from context + previous state
  Result: decoder always has access to the full input,
          weighted by relevance to the current output word
```

# Transformers — Study Notes

---

## 1. Why Transformers?

RNNs (even with Attention) process a sequence **one word at a time**:

```
x⟨1⟩ → x⟨2⟩ → x⟨3⟩ → ... → x⟨Tx⟩     (strictly sequential)
```

| Problem with RNN + Attention | Consequence |
|---|---|
| Sequential computation | Can't parallelise across time steps → slow training |
| Long dependency chains | Information can still degrade over very long sequences |
| One hidden state carries everything forward | Bottleneck even with attention helping |

**Transformers** remove recurrence entirely. Every word attends to every other word **simultaneously**, so the whole sequence can be processed in parallel.

> Core idea: **Attention + CNN-style parallelism.** Take the attention mechanism (which already lets a word "look at" other words) and apply it in a fully parallel, non-recurrent architecture — self-attention + multi-head attention.

```
RNN + Attention:   process word-by-word, attend to encoder states
Transformer:       process all words at once, attend to each other directly
```

---

## 2. Self-Attention — Core Intuition

Self-attention computes a new, **context-aware representation** for every word by letting it attend to all other words in the same sentence (including itself).

For each word we compute three vectors (learned via weight matrices):

```
q = query   →  "what am I looking for?"
k = key     →  "what do I contain / advertise?"
v = value   →  "what information do I actually give if attended to?"

q⟨t⟩ = Wq · x⟨t⟩
k⟨t⟩ = Wk · x⟨t⟩
v⟨t⟩ = Wv · x⟨t⟩
```

**A(q, K, V)** — the attention-based vector representation of a word — is computed for **every word** in the sentence using its query against all keys/values.

> Analogy: think of it like a search engine. `q` is your search query, `k` is the index keys of every document, and `v` is the actual document content returned. You match your query against every key, then pull a weighted blend of the values.

---

## 3. Scaled Dot-Product Attention

```
Attention(Q, K, V) = softmax( Q·Kᵀ / √dk ) · V
```

| Step | What happens |
|---|---|
| 1. Score | `Q · Kᵀ` — dot product of query with every key → similarity score |
| 2. Scale | Divide by `√dk` (dk = key dimension) → keeps gradients stable |
| 3. Normalise | `softmax(...)` → converts scores into weights that sum to 1 |
| 4. Weighted sum | Multiply weights by `V` → final attention output for that word |

```
        word 1   word 2   word 3   ...
q⟨1⟩ · [ k⟨1⟩     k⟨2⟩     k⟨3⟩   ... ]  →  softmax  →  weights
                                                          ↓
A⟨1⟩ = Σ_t  weight(1,t) · v⟨t⟩
```

Every word's output `A⟨t⟩` is a blend of **all** words' values, weighted by how relevant each word is to word `t`.

---

## 4. Multi-Head Attention

Instead of computing attention once, do it **multiple times in parallel** with different learned projections of Q, K, V — each "head" can learn to focus on a different kind of relationship (e.g., syntax, coreference, adjacency).

```
head_i = Attention(Q·Wq_i , K·Wk_i , V·Wv_i)     for i = 1 ... h

MultiHead(Q,K,V) = Concat(head_1, ..., head_h) · Wo
```

| Single-Head Attention | Multi-Head Attention |
|---|---|
| One "view" of relationships between words | Multiple parallel "views" (e.g. one head tracks grammar, another tracks meaning) |
| Cheaper | Richer representation, still fully parallelisable |

> Each head has its own `Wq, Wk, Wv` — they're learned independently and concatenated at the end, then passed through one more linear layer `Wo`.

---

## 5. Positional Encoding

Self-attention has **no notion of word order** — it treats the sentence like a set, not a sequence. So position information is injected by **adding** a positional encoding vector to each word's embedding before it enters the model:

```
Input to model = Word Embedding + Positional Encoding

PE(pos, 2i)   = sin( pos / 10000^(2i/d) )
PE(pos, 2i+1) = cos( pos / 10000^(2i/d) )
```

This gives every position a unique, fixed pattern that the model can learn to use as a signal for relative/absolute position.

---

## 6. Encoder Block

```
Input Embeddings + Positional Encoding
              ↓
     Multi-Head Self-Attention
              ↓
        Add & Norm   (residual connection + layer norm)
              ↓
     Feed-Forward Network (2 linear layers, ReLU in between)
              ↓
        Add & Norm
              ↓
        → next Encoder block (stacked N times)
```

- **Add & Norm**: residual connection (`x + Sublayer(x)`) followed by layer normalisation — stabilises training and preserves gradient flow through deep stacks.
- **Feed-Forward Network**: applied independently to each position, adds non-linearity/capacity beyond attention.

---

## 7. Decoder Block

The decoder is similar but has **two** attention sub-layers instead of one:

```
Output Embeddings (shifted right) + Positional Encoding
              ↓
     Masked Multi-Head Self-Attention      (can't look at future words)
              ↓
        Add & Norm
              ↓
     Multi-Head Attention (Encoder-Decoder)  (Q from decoder, K & V from encoder)
              ↓
        Add & Norm
              ↓
     Feed-Forward Network
              ↓
        Add & Norm
              ↓
        → next Decoder block (stacked N times)
              ↓
      Linear + Softmax → output word probabilities
```

| Sub-layer | Purpose |
|---|---|
| Masked self-attention | Each output word can only attend to previous output words — preserves autoregressive generation |
| Encoder-decoder attention | Lets the decoder attend to the **entire input sentence**, just like classic attention |
| Feed-forward | Same role as in the encoder |

> The mask is applied by setting future positions' attention scores to `-∞` before the softmax, so their weight becomes ~0.

---

## 8. Full Architecture at a Glance

```
Inputs                                  Outputs (shifted right)
  ↓                                             ↓
Embedding + Positional Encoding        Embedding + Positional Encoding
  ↓                                             ↓
┌─────────────┐                        ┌───────────────────────┐
│  ENCODER ×N │ ── K, V (from top) ──▶ │  DECODER ×N            │
│  Self-Attn  │                        │  Masked Self-Attn      │
│  Add & Norm │                        │  Add & Norm             │
│  Feed Fwd   │                        │  Enc-Dec Attn (Q here)  │
│  Add & Norm │                        │  Add & Norm             │
└─────────────┘                        │  Feed Fwd               │
                                        │  Add & Norm             │
                                        └───────────────────────┘
                                                 ↓
                                          Linear + Softmax
                                                 ↓
                                          Output probabilities
```

---

## 9. Why Transformers Win

| Aspect | RNN / RNN + Attention | Transformer |
|---|---|---|
| Computation | Sequential (word by word) | Fully parallel across all positions |
| Long-range dependencies | Degrades over distance (even with attention, one path per pair) | Direct connection between any two words |
| Training speed | Slow (can't parallelise time steps) | Fast (GPU/TPU friendly, matrix multiplies) |
| Position awareness | Implicit (via recurrence) | Explicit (positional encoding, since attention alone is order-agnostic) |
| Interpretability | Attention heatmap (1 view) | Multi-head attention gives several complementary views |

---

## 10. Summary

```
Self-Attention:
  Every word computes q, k, v
  A(q,K,V) = softmax(Q·Kᵀ / √dk) · V
  → context-aware representation using every other word

Multi-Head Attention:
  Run self-attention h times in parallel (different learned projections)
  Concatenate + linear projection → richer representation

Positional Encoding:
  Injects word-order information (sin/cos) since attention is order-agnostic

Encoder:
  (Self-Attn → Add&Norm → FeedForward → Add&Norm) × N

Decoder:
  (Masked Self-Attn → Add&Norm → Enc-Dec Attn → Add&Norm → FeedForward → Add&Norm) × N
  → Linear → Softmax → output word

Result: a fully parallelisable architecture that models relationships
        between all words directly, replacing recurrence with attention.
```
