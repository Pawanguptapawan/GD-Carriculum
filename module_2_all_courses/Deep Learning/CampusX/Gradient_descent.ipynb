# Batch Gradient Descent vs Stochastic Gradient Descent

## ⚡ Speed per Epoch
- **Batch Gradient Descent (BGD):**
  - Faster per epoch
  - Processes the entire dataset at once

- **Stochastic Gradient Descent (SGD):**
  - Slower per epoch
  - Processes one example at a time

---

## 📉 Convergence Behavior
- **BGD:**
  - Slower convergence
  - Updates parameters **once per epoch**
  - Fewer updates → slower movement toward minima

- **SGD:**
  - Faster convergence
  - Updates parameters **after every example**
  - More updates → faster movement toward minima

---

## 📊 Loss Curve
- **BGD:**
  - Smooth loss curve
  - Uses the same full dataset for every update
  - Less noisy

- **SGD:**
  - Noisy loss curve
  - Uses different data points for each update
  - More noisy due to frequent updates

---

## 🎯 Convergence Accuracy
- **BGD:**
  - More precise convergence
  - Moves steadily toward the minimum

- **SGD:**
  - Less precise
  - May oscillate around the minimum
  - Might not settle exactly at the minimum

---

## 🧗 Escaping Local Minima
- **BGD:**
  - Hard to escape local minima
  - Small, consistent steps

- **SGD:**
  - Can escape local minima
  - Random, sometimes large jumps

---

## ⚙️ Computational Efficiency
- **BGD:**
  - Uses vectorization on the full dataset
  - Efficient for smaller datasets

- **SGD:**
  - Processes one sample at a time
  - Less efficient per update but more flexible

---
