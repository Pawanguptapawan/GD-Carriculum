# Optimization Technique: Batch Normalization

## 📌 What is Batch Normalization?

Batch Normalization (BatchNorm) is a technique that:

- Normalizes the **activation values (Z)** of hidden layers
- Uses the **mean (μ)** and **variance (σ²)** of the current mini-batch
- Is applied **before or after the activation function**

---

## 🧠 Why Do We Need It?

- Unnormalized data → ❌ slow training  
- Unstable distributions → ❌ poor convergence  

After normalization:
- ✅ Faster training  
- ✅ More stable convergence  
- ✅ Better optimization landscape  

---

## 📉 Covariate Shift

### 🔹 Covariate Shift
- Occurs when:
  - Input distribution \( P(X) \) changes
  - But \( P(Y|X) \) remains the same
- Causes poor performance on new data

---

### 🔹 Internal Covariate Shift

- Change in distribution of **layer activations** during training
- Happens because:
  - Weights keep updating
  - Each layer sees constantly changing inputs

#### ⚠️ Effects:
- Slower convergence  
- Requires lower learning rates  
- Makes deep networks harder to train  

---

## ✅ Batch Normalization Solves This

- Keeps input distribution of each layer **stable**
- Forces:
  $$
  \mu = 0,\quad \sigma = 1
  $$

---

## ⚙️ BatchNorm Computation

### Step 1: Normalize

$$
Z_{\text{norm}} = \frac{Z - \mu}{\sigma}
$$

---

### Step 2: Scale and Shift

$$
Z_{\text{BN}} = \gamma Z_{\text{norm}} + \beta
$$

- \( \gamma \) → scale (learnable)
- \( \beta \) → shift (learnable)

---

### Step 3: Activation

$$
A = g(Z_{\text{BN}})
$$

---

## 🔁 Full Flow

```
Z → Z_norm → Z_BN → Activation → A
```

- Each neuron has its own:
  - \( \gamma \)
  - \( \beta \)

---

## 🧪 Training vs Testing

### 🔹 During Training
- Mean \( \mu \) and variance \( \sigma \) are computed from the **current batch**

---

### 🔹 During Testing

- Only one sample is available
- So we use **running averages** (estimated during training)

👉 Computed using **Exponentially Weighted Moving Average (EWMA)**

---

## 📦 Parameters Stored per Neuron

- Learnable:
  - \( \gamma \)
  - \( \beta \)

- Non-learnable (tracked during training):
  - Running mean \( \mu \)
  - Running variance \( \sigma^2 \)

---

## 🚀 Summary

- BatchNorm normalizes hidden layer activations  
- Reduces **internal covariate shift**  
- Speeds up training  
- Improves stability  
- Enables higher learning rates
