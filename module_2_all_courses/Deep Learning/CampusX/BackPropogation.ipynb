# Backpropagation Algorithm (Single Hidden Layer)

We consider a neural network with:

- **Input layer:** `x1, x2`  
- **Hidden layer:** `y3, y4`  
- **Output layer:** `y5`  

---

## 1) Prediction Error

For the output neuron `y5`:

$$
\text{Error} = y_{\text{target}} - y_5
$$

---

## 2) Gradient-Based Weight Update Rule

The general update rule for weights is:

$$
\Delta w_{ij} = \eta \cdot \delta_j \cdot o_i
$$

Where:

- `η` = learning rate  
- `δj` = local gradient (error term) at neuron `j`  
- `oi` = output of neuron `i` (input to neuron `j`)  

### Weight update step:

$$
w_{ij} \leftarrow w_{ij} + \Delta w_{ij}
$$

---

## 3) Output Layer Delta

For neuron `5` (sigmoid activation):

$$
\delta_5 = y_5 (1 - y_5)(y_{\text{target}} - y_5)
$$

---

## 4) Hidden Layer Deltas

### For neuron `3`:

$$
\delta_3 = y_3 (1 - y_3) \cdot (w_{3,5} \cdot \delta_5)
$$

### For neuron `4`:

$$
\delta_4 = y_4 (1 - y_4) \cdot (w_{4,5} \cdot \delta_5)
$$

### General hidden layer formula:

$$
\delta_h = y_h (1 - y_h) \sum_k w_{h,k} \delta_k
$$

---

## 5) Example Weight Updates

### Hidden → Output (Neuron 4 to 5)

$$
\Delta w_{4,5} = \eta \cdot \delta_5 \cdot y_4
$$

$$
w_{4,5} \leftarrow w_{4,5} + \Delta w_{4,5}
$$

---

### Input → Hidden (Neuron 1 to 3)

$$
\Delta w_{1,3} = \eta \cdot \delta_3 \cdot x_1
$$

$$
w_{1,3} \leftarrow w_{1,3} + \Delta w_{1,3}
$$

---

### Input → Hidden (Neuron 2 to 4)

$$
\Delta w_{2,4} = \eta \cdot \delta_4 \cdot x_2
$$

$$
w_{2,4} \leftarrow w_{2,4} + \Delta w_{2,4}
$$

---

## Notes

- `δ` (delta) represents how much a neuron contributed to the error.
- Backpropagation works by:
  1. Computing error at the output
  2. Propagating it backward
  3. Updating weights accordingly

```python
import numpy as np

class NetworkNN:
    def __init__(self,input_size,hidden_size,output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        
        # Initialize weights and biases
        self.weights_input_hidden = np.random.randn(self.input_size, self.hidden_size)
        self.weights_hidden_output=np.random.randn(self.hidden_size,self.output_size)
        self.bias_hidden = np.zeros((1,self.hidden_size))
        self.bias_output = np.zeros((1,self.output_size))
    
    def sigmoid(self,x):
        return 1/(1+np.exp(-x))
    
    def sigmoid_derivative(self,x):
        return x*(1-x)

    
    def Forward(self,X):
        self.hidden_activation = np.dot(X,self.weights_input_hidden)+self.bias_hidden
        # (1,2),(2,4)->(1,4)+(1,4)

        self.hidden_output=self.sigmoid(self.hidden_activation)
        # (1,4)
        self.output_activation=np.dot(self.hidden_output,self.weights_hidden_output)+self.bias_output
        # (1,4),(4,1)->(1,1)+(1,1)
        self.predicted_output=self.sigmoid(self.output_activation)
        # (1,1)

        return self.predicted_output

    
    def backward(self, X, y, learning_rate):
        # Output layer
        # X=(1,2)
        # y=(1,1)
        output_error = y - self.predicted_output
        # (1,1)-(1,1)
        output_delta = output_error * self.sigmoid_derivative(self.predicted_output)
        # (1,1)

        # Hidden layer
        hidden_error = np.dot(output_delta, self.weights_hidden_output.T)
        #(1,1).(1,4)->(1,4)
        hidden_delta = hidden_error * self.sigmoid_derivative(self.hidden_output)
        # (1,4)

        # Update weights
        self.weights_hidden_output += np.dot(self.hidden_output.T, output_delta) * learning_rate
        # (4,1).(1,1)->(4,1)
        self.bias_output += np.sum(output_delta, axis=0, keepdims=True) * learning_rate
        # (1,1)

        self.weights_input_hidden += np.dot(X.T, hidden_delta) * learning_rate
        # (2,1). (1,4)->(2,4)
        self.bias_hidden += np.sum(hidden_delta, axis=0, keepdims=True) * learning_rate
        # (1,4)

        


    
    def train(self,X,y,epochs,learning_rate):
        for epoch in range(epochs):
            output=self.Forward(X)
            self.backward(X,y,learning_rate)
            if epoch%10==0:
                loss=np.mean((y-output)**2)
                print(f"Epoch {epoch}, Loss:{loss}")
    

    def print_shapes(self):
        print("Weights Input → Hidden:", self.weights_input_hidden.shape)
        print("Weights Hidden → Output:", self.weights_hidden_output.shape)
        print("Bias Hidden:", self.bias_hidden.shape)
        print("Bias Output:", self.bias_output.shape)
        

        
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

nn = NetworkNN(input_size=2, hidden_size=4, output_size=1)
nn.train(X, y, epochs=100, learning_rate=0.01)

output = nn.Forward(X)
print("Predictions after training:")
print(output)


nn.print_shapes()
```
