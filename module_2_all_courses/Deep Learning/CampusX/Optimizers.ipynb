# Gradient Descent with Momentum

## 📌 What is Momentum?

Gradient Descent with Momentum is an optimization algorithm that:

- Accelerates convergence
- Smooths parameter updates
- Helps escape saddle points and shallow local minima

---

## ⚙️ Update Rules

### Velocity Update

$$
V_{dW} = \beta \cdot V_{dW} + (1 - \beta) \cdot dW
$$

$$
V_{db} = \beta \cdot V_{db} + (1 - \beta) \cdot db
$$

---

### Parameter Update

$$
W = W - \alpha \cdot V_{dW}
$$

$$
b = b - \alpha \cdot V_{db}
$$

---

## 🔢 Notation

- \( V_{dW}, V_{db} \): Velocity (momentum term)  
- \( \beta \): Momentum coefficient (typically ~0.9)  
- \( dW, db \): Gradients  
- \( \alpha \): Learning rate  

---

## 🧠 Intuition

- Momentum accumulates past gradients → builds **inertia**
- Reduces oscillations in steep directions
- Speeds up movement in consistent directions

---

## 📉 Behavior on Contour Plots

- **Vertical direction (steep slope):**
  - Want slower movement
  - Momentum smooths oscillations

- **Horizontal direction (flat slope):**
  - Want faster movement
  - Momentum accelerates progress

---

## ⚡ Physics Analogy

- \( dW, db \) → **Acceleration**  
- \( V_{dW}, V_{db} \) → **Velocity**  
- \( \beta \) → **Friction (resistance)**  

👉 Just like a ball rolling down a hill:
- Builds speed over time
- Doesn’t zig-zag too much

---

## 🎯 Advantages

- Faster convergence  
- Reduced oscillations  
- Better performance in ravines  
- Helps escape saddle points  

---

## ⚙️ Hyperparameters

- \( \alpha \) → learning rate  
- \( \beta \) → momentum term  

---

## 🚀 Summary

- Momentum = **smoothed gradient descent**  
- Uses past gradients to guide updates  
- Improves speed + stability of training
