# Regularization (L1 & L2)

## 📌 L2 Regularization (Ridge)

Cost function:

$$
C = L + \frac{\lambda}{2n} \sum W^2
$$

- Adds squared weights penalty
- Encourages **small but non-zero weights**
- Helps reduce overfitting

---

## 📌 L1 Regularization (Lasso)

Cost function:

$$
C = L + \frac{\lambda}{2n} \sum |W|
$$

- Adds absolute weights penalty
- Encourages **sparse weights (many become 0)**
- Useful for feature selection

---

## 🧠 How Regularization Reduces Weights

Updated loss function:

$$
L' = L + \frac{\lambda}{2n} \sum W^2
$$

---

### 🔹 Gradient of New Loss

$$
\frac{dL'}{dW} = \frac{dL}{dW} + \lambda W
$$

---

### 🔹 Weight Update Rule

$$
W = W - \eta \left( \frac{dL}{dW} + \lambda W \right)
$$

---

## ⚡ Intuition

- The term **\( \lambda W \)** pulls weights toward **zero**
- Larger weights → larger penalty → stronger reduction
- Prevents the model from relying too much on any single feature

---

## 🔍 Key Differences

| Feature            | L1 Regularization | L2 Regularization |
|--------------------|------------------|------------------|
| Penalty            | \|W\|            | W²               |
| Effect             | Sparse weights   | Small weights    |
| Feature selection  | Yes              | No               |
| Stability          | Less stable      | More stable      |

---

## 🚀 Summary

- **L1** → drives weights exactly to **0** (sparse model)  
- **L2** → shrinks weights **close to 0** (smooth model)
