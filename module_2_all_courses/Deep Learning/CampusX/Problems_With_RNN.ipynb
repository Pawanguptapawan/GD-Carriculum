# LSTM

```python
# we maintain 2 type of context: long term(cell state) and short term(hidden_state)
# in input we take 3 context : Prev cell state , prev hidden state , input fro the current state
# as output we got : current hidden state , current cell state
# in processing state : we update the cell state , calculate hidden state
# in the update cell state based on current context we decide which term we have to remove and which term we have to add. 
```
