```python
# Exponentially weighted averages:
    # The Exponentially Weighted Moving Average (EWMA) is a statistical method used to analyze time-series data by calculating a moving average that assigns exponentially decreasing weights to older observations, giving higher importance to recent data. It is used in time series analysis, finance, and AI (e.g., Optimizers like Adam) to smooth data and detect trends by balancing current data with past information using a smoothing factor B, 

    #The average at time t(vt) is v[t]=B.v[t-1]+(1-B).theta[t]  where theta is the current value and B is a hyperparameter between 0 and 1 (commonly 0.9).
    # 1/(B-1) if B=0.98
    # 1/(B-1) ~ 50
    # if the B is high  the plot is much smoother 
    # if B is very small then  we have much shorter window so plot will be much more noisy and much more susceptible to outliers.  but it adapts much more quickly to how values are changes.

```

```python
# Gradient Descent with momentum : let's say we are trying to optimize a cost function. gradient (mini batch or batch) slowly oscillate towards the minimum and this up and down oscillations slows down gradient descent and prevent us from using a much larger learning rate . if the learning rate is much larger  then we might end up  overshoot and end up diverging like so. so the need to prevent oscillations from getting too big forces we need to use a LR that's not itself too large.

# we want that the learning to be slower because we don't want those oscillations . In the contour plot on the vertical axis we need slower learning(small jumps) and on horizontal axis we need faster learning (less number of jumps).
# if we use gradient descent with momentum : during iteration t we would compute the usual derivatives dw and db on current mini batch. ( this will work very well  on batch gradient descent also).
# V[dw]=B*V[dw]+(1-B)*dw
# W = W-alpha*V[dw]
# b= b-alpha*V[db]
# if we average out these gradients we found that the oscillations in the vertical direction will tend to average out to somthing closer to zero. in vertical direction we want to slow the things donw , this will average out positive and negative numbers so the average will be close to zero. on the horizontal directions all the derivative pointing to the right of the horizontal direction, so the average in the horizontal direction will still be preety big. 

# V[dw]=B*V[dw]+(1-B)*dw   this derivative terms dw and db providing acceleration. 
# V[dw] and V[db] we can think as velocity. 
# B we can think as friction and it prevents ball from speeding up without limit.

# On iteration t:
    # compute dw,db on the current mini batch
        # V[dw]=B*V[dw]+(1-B)*dw
        

# alpha and B is the hyperparameter. 

# we uaually don't use bias correction with momentum because after 10 iterations moving average will have warmed up and is no longer a bias estimate. 

# gradient descent with momentum often we use term ommited(1-B) so we end up with 
# V[dw] = B*V[dw]+dW
# the net effect of using this version is that v[dW] end up being scaled by a factor of (1-B) so when we perform 
# W = W-alpha*V[dw]
# b= b-alpha*V[db].    these steps alpha just nedds to change by a corresponding value of 1/(1-B). It just affect what is the best value of the learning rate alpha. 
# But this formula is less intutive because if we end up tuning the hyparameter B then this affect scaling of v[dW] and v[dB] as well. so we need to retune the learning rate, alpha as well. 


# V[dw]=B*V[dw]+(1-B)*dw
# W = W-alpha*V[dw]
# b= b-alpha*V[db].  this is the best way. 

```

```python
# RMSProp :
    # problem is same in horzontal direction we want fast learning (less number of steps) and in vertical direction we want smaller learning (smaller jumps not higher jumps)

    # lets represent the vertical direction with b and horizontal direction with w.
    # on iteration t it compute dW,db on current mini batch. 
    #  S[dW]= B*S[dW]+(1-B)*(dW)**2
    # S[dB]=B*S[db]+(1-B)*(db)**2
    # W = W-(alpha*dW)/sqrt(S[dW])
    # b = b-(alpha*db)/sqrt(S[db]) 
    
# intution : we want to slow down all the oscillations into the vertical direction so with these terms S[dW] and S[db]. we are hoping that S[dW] will be relatively small so that we are dividing by relatively small number  (alpha*dW)/sqrt(S[dW]).  this term becomes very large and W will decrease with very high value. 
# and S[db] relatively large, (alpha*db)/sqrt(S[db])  will be small and b will decrease but with small value.   so this slow down the updates on a vertical dimension. 

# so these derivative becomes very large in vertical direction in compare to horizontal direction  so the slope will be very high in the direction of b. db is very large and dw is very small relatively. 
# db**2 will be relatively large so S[db] relativelty  large and dW is smaller so that S[dw] would be smaller. 
# updates in vertical direction is divide by very large number and updates in horizontal direction is divide by very small number so that helps damp out the oscillations.
# one effect of this is also that we can use larger learning rate and get faster learning without diverging in the vertical direction.  

```
