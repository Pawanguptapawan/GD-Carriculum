# Improving Deep Neural Networks — Study Notes

---

## 1. Bias and Variance Tradeoff

| Concept | Cause | Symptom |
|---------|-------|---------|
| **High Bias** (Underfitting) | Overly simple model; misses patterns | High training error **and** high test error |
| **High Variance** (Overfitting) | Model memorises noise in training data | Low training error but **high test error** |

### Fixing High Bias
- Increase model complexity (more layers, more units, polynomial features).
- Use a more powerful algorithm.
- Decrease the regularisation term.

### Fixing High Variance
- Simplify the model (fewer layers/features).
- Add more training data.
- Increase the regularisation term.

---

## 2. Regularisation
Regularization is a machine learning technique used to prevent overfitting by adding a penalty to model complexity, which encourages simpler, more generalizable models. It ensures the algorithm learns underlying trends rather than memorizing noise in the training data.

### L2 Regularisation (Weight Decay)

> J(w, b) = (1/m) · Σ Loss(y, ŷ) + (λ/2m) · ‖W‖²

- **‖W‖²** = `np.dot(W.T, W)` — Euclidean (Frobenius) norm.
- Gradient update with L2:

```
dJ/dW  ←  dJ/dW + (λ/m) · W
W      ←  W − α · (dJ/dW + (λ/m) · W)
       =  (1 − α·λ/m) · W − α · dJ/dW
```

> This factor `(1 − α·λ/m)` is why L2 is called **weight decay** — it shrinks weights at every step.

### L1 Regularisation

> J(w, b) = (1/m) · Σ Loss(y, ŷ) + (λ/m) · ‖W‖₁

- **‖W‖₁** = `np.sum(np.abs(W))` — Manhattan norm.
- Tends to produce sparse weights.

### How Regularisation Reduces Overfitting

| λ value | Effect |
|---------|--------|
| Very large | Weights ≈ 0 → network nearly linear → **underfits** |
| Very small | Regularisation has no effect → **overfits** |
| Just right | Weights small but non-zero → good generalisation |

> When λ is large, z = W·x + b is small, activations (e.g., tanh) stay in the linear regime, making the whole network roughly linear.

---

## 3. Dropout Regularisation

Randomly drop nodes in each layer for each training iteration.

```python
keep_prob = 0.8                                      # probability of keeping a node
d3 = np.random.rand(a3.shape[0], a3.shape[1]) < keep_prob
a3 = np.multiply(a3, d3)                             # zero out dropped nodes
a3 /= keep_prob                                      # inverted dropout: scale up
```

> **Key rule:** Dropout is used **only during training**, never at test time.

**Intuition:** The network cannot rely on any single feature, so it is forced to spread weights across many inputs — effectively shrinking the squared norm of the weights, similar to L2.

**Analogy:** Similar to Random Forests — each training step uses a different "sub-network", like training many sub-trees.

---

## 4. Other Regularisation Strategies

- **Data Augmentation** — Flipping, cropping, rotating images cheaply generates more training examples.
- **Early Stopping** — Stop training when the validation loss starts increasing. Mixes optimisation and variance reduction (a tradeoff of **orthogonalisation**).

> **Orthogonalisation principle:** It is cleaner to keep optimisation (reduce J) and generalisation (reduce overfitting) as separate concerns. Early stopping couples both, which makes tuning harder.

---

## 5. Input Normalisation

Scale all input features to a similar range before training.

```python
mu    = np.mean(X, axis=0)
sigma = np.std(X, axis=0)
X     = (X - mu) / sigma
```

> Apply the **same** μ and σ (computed on training data) to the test set.

**Why it matters:** Un-normalised features create elongated cost-function contours, forcing gradient descent to zigzag and converge slowly.

---

## 6. Vanishing & Exploding Gradients

| Problem | What happens | Typical cause |
|---------|-------------|---------------|
| **Vanishing gradients** | Gradients shrink layer-by-layer → early layers learn nothing | Sigmoid / tanh activations in very deep nets |
| **Exploding gradients** | Gradients grow uncontrollably → training diverges | Unbounded activations (ReLU in some configurations) |

### Weight Initialisation (Partial Fix)

Keep initial weights neither too large nor too small:

```python
# He initialisation (recommended for ReLU)
W = np.random.randn(n[l], n[l-1]) * np.sqrt(2 / n[l-1])

# Xavier initialisation (recommended for tanh)
W = np.random.randn(n[l], n[l-1]) * np.sqrt(1 / n[l-1])

# Yoshida variant
W = np.random.randn(n[l], n[l-1]) * np.sqrt(2 / (n[l] + n[l-1]))
```

## 7. Numerical Approximation of Gradients

Used to **verify** backprop implementations via the finite-difference method:

> dθᵢ ≈ [J(…, θᵢ + ε, …) − J(…, θᵢ − ε, …)] / (2ε)

Two-sided (centred) difference is more accurate than one-sided. This is computationally expensive — use only for debugging.

---

## 8. Gradient Checking

Compare analytical gradients (backprop) with numerical approximations:

```
for each i:
    dθ_approx[i] = [J(θ + ε·eᵢ) − J(θ − ε·eᵢ)] / (2ε)

check:  ‖dθ_approx − dθ‖ / (‖dθ_approx‖ + ‖dθ‖)  ≈ 10⁻⁷ (good)  /  10⁻⁵ (warning)
```

**Rules:**
- Use for **debugging only** — never during actual training.
- Include the regularisation term in J.
- Turn off Dropout before running gradient check.
- Run at random initialisation and again after a few training steps.

## 9. Optimisation Algorithms

### Mini-Batch Gradient Descent

Splits training data into small batches; updates parameters after each batch.

| Batch size | Behaviour |
|-----------|-----------|
| = m (full batch) | Slow per step, smooth convergence |
| = 1 (stochastic GD) | Fast but noisy updates |
| 32–512 (mini-batch) | Best of both worlds ✅ |

> Choose batch size as a **power of 2** (32, 64, 128, 256, 512) to exploit GPU memory alignment.

---

### Exponentially Weighted Moving Average (EWMA)

> V_t = β · V_{t−1} + (1 − β) · θ_t

Approximates the average over roughly **1/(1−β)** past values.

---

### Bias Correction

EWMA is initialised at 0, biasing early estimates downward. Correct with:

> V̂_t = V_t / (1 − β^t)

---

### Gradient Descent with Momentum

Smooth out oscillations by accumulating a velocity term:

```
V[dW] = β · V[dW] + (1 − β) · dW
V[db] = β · V[db] + (1 − β) · db
W     = W − α · V[dW]
b     = b − α · V[db]
```

**Physics analogy:** dW/db = acceleration, V[dW]/V[db] = velocity, β = friction.  
Hyperparameters: **α** (learning rate), **β** (momentum, default ≈ 0.9).

---

### RMSProp

Adapts learning rate per parameter by dividing by a moving average of squared gradients:

```
S[dW] = β · S[dW] + (1 − β) · dW²
S[db] = β · S[db] + (1 − β) · db²
W     = W − α · dW / (√S[dW] + ε)
b     = b − α · db / (√S[db] + ε)
```

> Dividing by large S dampens parameters with large gradients; small S speeds up parameters with small gradients.

---

### Adam (Adaptive Moment Estimation)

Combines **Momentum** + **RMSProp**, with bias correction:

```python
# Step 1: Compute moving averages
V[dW] = β1·V[dW] + (1−β1)·dW       # momentum
V[db] = β1·V[db] + (1−β1)·db
S[dW] = β2·S[dW] + (1−β2)·dW²      # RMSProp
S[db] = β2·S[db] + (1−β2)·db²

# Step 2: Bias correction
V̂[dW] = V[dW] / (1 − β1^t)
V̂[db] = V[db] / (1 − β1^t)
Ŝ[dW] = S[dW] / (1 − β2^t)
Ŝ[db] = S[db] / (1 − β2^t)

# Step 3: Update
W = W − α · V̂[dW] / (√Ŝ[dW] + ε)
b = b − α · V̂[db] / (√Ŝ[db] + ε)
```

| Hyperparameter | Typical default |
|----------------|----------------|
| α (learning rate) | tune this |
| β1 (momentum) | 0.9 |
| β2 (RMSProp) | 0.999 |
| ε (stability) | 10⁻⁸ |

## 10. Learning Rate Decay

Reduce α over training to take smaller steps near convergence.

| Method | Formula |
|--------|---------|
| Time-based | α = α₀ / (1 + decay_rate · epoch) |
| Step decay | α = α₀ · drop_rate^⌊epoch / epochs_drop⌋ |
| Exponential | α = α₀ · e^(−decay_rate · epoch) |

---

## 11. Saddle Points & Plateaus

In high-dimensional spaces, the gradient is rarely zero in all dimensions simultaneously — true local minima are rare. The bigger problem is **plateaus**, where gradients are near zero for a long stretch, slowing learning. Algorithms like Adam handle this better than vanilla gradient descent.

## 12. Hyperparameter Tuning

### Priority Order (Most → Least Important)
1. Learning rate **α**
2. Mini-batch size, β (momentum), number of hidden units
3. Number of layers, learning rate decay

### Search Strategies

| Strategy | Description |
|----------|-------------|
| **Grid Search** | Exhaustive; slow for many parameters |
| **Random Search** | Samples randomly; often more efficient |
| **Bayesian Optimisation** | Uses past results to pick next trial intelligently |
| **Successive Halving** | Tournament-style; eliminates poor configs early |
| **Manual Tuning** | Experience-driven, good for initial experiments |

> **Use appropriate scales:** For α, search on a log scale (e.g., 10⁻⁴ to 1), not linearly.

### Two Training Approaches

- **Babysitting one model** — Monitor and adjust a single model (when compute is limited).
- **Training many in parallel** — Train multiple configurations simultaneously and compare.

## 13. Batch Normalisation

Normalises the inputs to each layer, not just the input layer.

### Forward Pass

```python
mu      = np.mean(Z, axis=0)
sigma   = np.std(Z, axis=0)
Z_norm  = (Z − mu) / (sigma + ε)
Z_tilde = γ · Z_norm + β           # γ and β are learnable parameters
```

> When using batch norm, the bias parameter **b** of the linear step cancels out during normalisation and can be dropped. β (in BN) takes its role.

### Parameter Updates

```
W     = W     − α · dW
γ     = γ     − α · dγ
β_BN  = β_BN  − α · dβ_BN
```

### Why Batch Norm Works

1. **Reduces internal covariate shift** — Keeps the distribution of each layer's input stable, so later layers don't have to constantly re-adapt.
2. **Speeds up training** — Normalised inputs lead to better-conditioned gradients.
3. **Slight regularisation effect** — Mini-batch statistics add noise to Z[l], similar to Dropout. Using larger batches reduces this noise (and reduces the regularisation effect).

### Batch Norm at Test Time

At inference we may have only one example — mini-batch statistics are meaningless. Instead, use **exponentially weighted averages** of μ and σ² accumulated across training mini-batches:

```python
# During training (for layer l, mini-batch t):
# Track running_mu[l] and running_sigma2[l] via EWMA

# At test time:
Z_norm = (Z − running_mu[l]) / np.sqrt(running_sigma2[l] + ε)
Z_tilde = γ · Z_norm + β
```

## 14. Softmax Regression

Generalisation of logistic regression to **multi-class** classification (C classes):

> P(y = k | x) = e^(z_k) / Σⱼ e^(z_j)

Output layer has C units with a softmax activation; each unit gives the probability of one class.

## 15. Gradient Tape (TensorFlow / Autodiff)

> **GradientTape** records the sequence of operations during the forward pass. On the backward pass it replays them in reverse to compute gradients automatically.

```python
with tf.GradientTape() as tape:
    y_pred = model(x)
    loss   = loss_fn(y_true, y_pred)

gradients = tape.gradient(loss, model.trainable_variables)
```
