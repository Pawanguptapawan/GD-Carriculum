```python
# reinforcement learning:
# return in reinforcement learning: sum of (reward*discount_factor**iteration_number)
# discount factor : gamma

# Return = r1*(gamma**0)+r2*(gamma**1)+r3*(gamma*2)......


# making decisions: policies in reinforcement learning
# we always try to go to nearest reward.
#  another way we can choose action is to always go for larger reward or smaller reward.
# in this learning technique we use a function PI which maps states to the actions to take.
# this function is called policy. 
# goal of learning is to find the  policy to maximize the reward.

# key features of learning:  states,set of actions,rewards,discound factor gamma.return,policy PI
# marcov decision process(MDP) : the future only depends on current state and not on anything that might occured prior to getting the current state.
# MDP future is only depend on where you are not on how you get here.
# a agent choose an action a anf based on this action something happen in environemt. this action choose based on policy. that help to change state of agent by gaining a reward.
```

```python
# Bellman equation:
# help to calculate the state value function
# s:current_state
# R(s):reward of current state.
# a: current action
# s' : state you get to after taking action a
# a': action that you take in s'
#  Q(s,a)=R(s)+gamma * max(Q(s',a'))       | max on action
# for stochastic environment 
# Bellman Equation is:
# Q(s,a)=R(s)+gamma * E[max(Q(s',a')) max on action]. 
# here E is expected or average value




# discrete v/s continuous state in reinforcement learning:
# A discrete state space means:
# The agent can be in a finite or countable number of states
# States are clearly separated
#You can list them
# A continuous state space means:
# State is represented by real numbers
# Infinite possible values
# Cannot list all states



# epsilon greedy policy: 
# How to choose actions while still learing:
# option 1:
# pick action a that maximize Q(s,a)

# option 2:
# with prob 0.95 pick the action that maximize Q(s,a) (greedy action / Exploitatio)
# with  prob 0.05 , pick an action randomly. (exploration)

# option 2 approach known as epsilon greedy policy here epsilon is the probability to take a random action.
# initally epsilon is very very high because we starts with some random actions and gradually value of epsilon get decreased.


# mini batch and soft update to tunning the algorithm:
# divide the entire dataset into mini  batches and provide as input to network. it takes less time to learning. 
# every iter takes a subset of data.
# Mini-batch gradient descent is an optimization algorithm that updates model parameters by calculating gradients on small subsets (typically 10–1000 examples) of training data, rather than the whole set or single points


# Stochastic Gradient Descent (SGD) is an iterative optimization algorithm used to minimize loss functions in machine learning by updating model parameters using small, randomly selected subsets (or single instances) of data rather than the entire dataset. 




# Batch Gradient Descent (BGD) is a machine learning optimization algorithm that calculates the gradient of the cost function by using the entire dataset for each update.
    




 
```
