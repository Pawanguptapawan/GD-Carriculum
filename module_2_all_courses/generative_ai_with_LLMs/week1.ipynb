```python
# Introduction to AWS labs

# three variance of the transformer model; encoder-only encoder-decoder models, and decode-only. Each of these is trained on a different objective, and so learns how to carry out different tasks. Encoder-only models are also known as Autoencoding models, and they are pre-trained using masked language modeling. Here, tokens in the input sequence or randomly mask, and the training objective is to predict the mask tokens in order to reconstruct the original sentence. This is also called a denoising objective. Autoencoding models spilled bi-directional representations of the input sequence, meaning that the model has an understanding of the full context of a token and not just of the words that come before. Encoder-only models are ideally suited to task that benefit from this bi-directional contexts.
# for example setiment analysis, name entity recognition, word classification
# examples: Bert

# decoder only or autoregressive models: decoder-only or autoregressive models, which are pre-trained using causal language modeling. Here, the training objective is to predict the next token based on the previous sequence of tokens. Predicting the next token is sometimes called full language modeling by researchers. Decoder-based autoregressive models, mask the input sequence and can only see the input tokens leading up to the token in question. The model has no knowledge of the end of the sentence. 
# examples: GPT   


# The final variation of the transformer model is the sequence-to-sequence model that uses both the encoder and decoder parts off the original transformer architecture. 
# s2s pretrain the encoder using span corruption, which mask random sequences of input tokens,those mask tokens are replaced with a setinel token(are special token addes to the vocabulary , but do not correspond to any actual word from the input text).the decoder is then tasked with reconstructing of mask token sequences auto-regressivly. the output of sentinel token followed bt the predict tokens. uses: translation,sumarization,and question-answering.

    
```

```python
# Fine tunning on a single task:

# good results can be achived with some good examples often 500-1000 can result in good result  in contrast of billion of pieces of text that the model saw during fine tunning. a potential downside to fine tunning on a single task the process may lead to catestropic forgetting. 
# catestropic forgetting: fine tuning can significantly increase the performance of a model on a specific task but can leas to reduction in ability on other tasks.
# a failure mode in neural networks where a models forgets previoulsy learned info upon training on new and different data.

# How to avoid catastropic forgetting: first note that you ight not have to.Fien tune on multi tasks at the same time.
# consider PEFT which preserves the weights of the original LLM  and train only a small number of task speicifc adapter layers and parameters. 
# One way to mitigate catastrophic forgetting is by using regularization techniques to limit the amount of change that can be made to the weights of the model during training.
# Catastrophic forgetting occurs when a machine learning model forgets previously learned information as it learns new information.
# Catastrophic forgetting is a common problem in machine learning, especially in deep learning models.

```

```python
# Model evaluation metrics:
# accuracy for deterministic model.
# non deterministic model -> rouge (used for text summarization) and bleu score(used for text translation)

# for only a specific task.

# for unigrams ( a single word) 
# rouge-1 recall = unigram matches / unigrams in human reference
# rouge-1 precision = unigram matches / unigrams in output
# rouge-1 F1 score = 2* (precison X recall)/(precision+recall)

# for bigrams ( a pair of words) 
# rouge-2 recall = unigram matches / unigrams in human reference
# rouge-2 precision = unigram matches / unigrams in output
# rouge-2 F1 score = 2* (precison X recall)/(precision+recall)


# for longest common subsequences(rouge-L)
# rouge-L recall = LCS(generated,reference) / unigrams in human reference
# rouge-L precision = LCS(gen,ref) / unigrams in output
# rouge-L F1 score = 2* (precison X recall)/(precision+recall


# for multitask rouge score is not comparable.

# BLEU ( Bilingual evaluation under study) score : score is calculated using the average precision over multiple n-gram sizes.
# BLEU matric: Avg(precision acorss range of n-gram sizes)



# Evaluation BenchMarks :GLUE (general langauge understaing evaluation) , SUPERGLUE , HELM
# for massive models : MMLU(massiva multitask language understanding), Big bench Hard 
# Holistic evaluation of Langauge Models(Helm) measuring 7 metrics (accuracy,calibration, robustness,fairness,bias,toxisity, efficiency)

 
```

```python
# PEFT focuses on updating only a small subset of parameters instead of training the entire model. Some PEFT techniques freeze most of the model’s weights and fine-tune only a limited set of existing parameters. They may also add a small number of new parameters or layers and train only these new components, making the process more efficient and less resource-intensive.
# PEFT tradeoffs-> memory efficiency,parameter effciciency , training speed, model performance, inference costs.
# PEFT Methods:
# 1.Selective (select subset of initial LLM parametrs to fine tune) 
# 2. Reparameterization also reduce the number of paramerrs to train by creating new low rank transformation of the orignal network weigths.( Reparameterization model weigths using a low rank representation like LORA) 
# 3. Additive carry out fine tunnig by keeping all of the original LLM weigths frozen and introducing new trainabel components(parameters to model like adapters method add trainable layers add new trainable layers to the architectire of the model typically insode the encoder or decoder components after the attention or feed-forward layers,   soft prompts methods keep the model architecture fixed and frozen and focus on manipulating the input of achieve better performance. this can be done by adding trainable parameters to the prompt embedding or keeping the input fixed and retraining the embedding weights.)



```

```python
# LoRA (low rank adaptation of LLM ) : 
# 1. Freeze most of the original LLM Weights.
# 2. Inject 2 rank decomposition matrices.
# 3. train the weights of the smaller matrices.

# steps:
# 1. Matrix multiply the low rank matrices.
# B * A = B X A
# 2. Add to original weights:
# updated parameters : Freeze parameters + B X A ( this is provided to the self attention layer)


# Example
# transformer weights have dimensions of dxk= 512 x 64 =  32,768 parameters.

# in loRa with rank r=8
# A has dim r x k = 8 x 64 = 512 parameters
# B has dim of d x k = 512 x 8 = 4096 parameters
# 86% reduction in parameters to train   so LoRa allows us to significantly reduce the number of trainable parametrs.

# 1. train different rank decomposition matrices for different tasks.
# 2. update wigths before inference.



```

```python
def print_number_of_trainable_model_parameters(model):
    trainable_params=0
    all_model_params=0
    for _,param in model.named_parameters():
        all_model_params+=param.numel()
        if param.requires_grad:
            trainable_params+=param.numel()
    return trainable_params,all_model_params

print(print_number_of_trainable_model_parameters(original_model))
```

```python
def tokenize_function(example):
    start_prompt = 'Summarize the following conversation.\n\n'
    end_prompt = '\n\nSummary: '
    prompt = [start_prompt + dialogue + end_prompt for dialogue in example["dialogue"]]
    example['input_ids'] = tokenizer(prompt, padding="max_length", truncation=True, return_tensors="pt").input_ids
    example['labels'] = tokenizer(example["summary"], padding="max_length", truncation=True, return_tensors="pt").input_ids
    
    return example
```

```python
tokenized_datasets=hugging_face_dataset.map(tokenize_function,batched=True)
tokenized_dataset=tokenized_datasets.remove_columns(['id','topic','dialogue','summary',])
```
